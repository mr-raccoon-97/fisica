### Espectros rotacionales y vibracionales de las moléculas.

En la aproximación de BO, la dinámica de los electrones genera un potencial que presenta un mínimo como función de la distancia internuclear.

En esta visión, podemos describir la dinámica molecular en término de las fluctuaciones de la distancia y rotaciones de la molécula.

Planteamos la ecuación de Schrödinger para los grados de libertad relativos de dos núcleos de masas $M_1$ y $M_2$, separados por una distancia $R$. Para describir el movimiento de los núcleos, es conveniente separar el movimiento del centro de masa y los movimientos relativos. Hamiltoniano sera entonces:

$$ \hat{H} = \hat{H}_{\text{CM}} + \hat{H}_{\text{relativo}} $$

Y la función de onda:

$$\ket{\psi} = \ket{\psi_{\text{CM}}} \otimes \ket{\psi_{\text{relativo}}}$$

Podemos escribir el Hamiltoniano relativo en términos de la coordenada de la distancia internuclear $R$:

$$ \hat{H}_{relativo} = -\frac{\hbar^2}{2\mu} \left( \frac{1}{R^2} \frac{\partial}{\partial R} \left( R^2 \frac{\partial}{\partial R} \right) - \frac{1}{R^2} \frac{\hat{L}^2}{\hbar^2} \right) + \hat{V}(R) $$


En donde $\mu = \frac{M_1 M_2}{M_1 + M_2}$ es la masa reducida, $R$, $\theta$ y $\phi$ son las coordenadas esféricas del vector internuclear $\vec{R}$ y $V(R)$ es el potencial efectivo que proviene de la dinamica electronica calculado para las configuraciones nucleares fijas.

Este hamiltoniano admite soluciones de la forma:

$$ \langle R, \theta, \phi \ket{\psi} = F(R) Y_{l}^{m}(\theta, \phi) $$

En donde $Y_{l}^{m}(\theta, \phi)$ son las funciones esféricas armónicas, las cuales son autovalores del operador $\hat{L}^2$ y $\hat{L}_z$.

$$ \hat{L}^2 \ket{Y_{l}^{m}} = l(l+1) \hbar^2 \ket{Y_{l}^{m}} $$

El problema puede ser reducido a una ecuación de schrödinger en una dimensión para la función radial $u(R) = R F(R)$.

$$ \left[ -\frac{\hbar^2}{2\mu} \frac{d^2}{d R^2} + \left( \frac{l(l+1)\hbar^2}{2\mu R^2} + V(R) \right) \right] u(R) = E u(R) $$

Hasta acá sé considero que el momento angular total de la molécula se debe a la rotación de los núcleos. Sin embargo, los electrones que conforman la ligadura molecular también contribuyen al momento angular total.



A partir de ahora supongamos el caso de momento angular electrónico nulo. El termino:

$$ U_l(R) \equiv \frac{l(l+1)\hbar^2}{2\mu R^2} + V(R) $$

Puede ser interpretado como un potencial efectivo, en el que la contribución centrífuga se suma a la energía potencial de la ligadura electronica. Luego, la ecuación de Schrödinger para la función radial $u(R)$ se reduce a:

$$ \left( -\frac{\hbar^2}{2\mu} \frac{d^2}{d R^2} + U_l(R) \right) u(R) = E u(R) $$

Con una energía potencial que depende del aspecto rotacional de la molécula. En la figura se puede observar el potencial efectivo para diferentes valores de $l$.



En la figura se puede observar que para un rango acotado de valores pequeños de momento angular, la forma del potencial efectivo cambia muy poco con respecto a $U_0$, sobre todo en la zona cercana al mínimo con $R \approx R_0$. 

Para pequeñas amplitudes del movimiento nuclear, la región de los potenciales efectivos cercanas a $R_0$, pueden ser consideradas como potenciales armónicos.

El efecto de levantar $l$, es el de disminuir la profundidad del pozo de potencial, por lo que se espera que la variación de $l$, no modifica el aspecto vibracional de la molécula. 

Se propone entonces la hipótesis **roto-vibracional**, en la que se considera que los espectros rotacionales y vibraciones de las moléculas son independientes para oscilaciones de baja amplitud.

Esto es, $R \rightarrow R_0$, por lo que se aproxima la energía rotacional como:

$$ E_{rot}^l \approx \frac{\hbar^2 l(l+1)}{2\mu R_0^2} $$

Luego la ecuación de Schrödinger para la vibración de la molécula se reduce a:

$$ \left( -\frac{\hbar^2}{2\mu} \frac{d^2}{d R^2} + V(R) \right) u(R) = (E - E_{rotacional}^l) u(R) = E_{vibracional}^l u(R) $$

Es decir, en esta aproximación el aspecto angular del estado de la molécula, no altera la dinámica radial, y ambos grados de libertad se consideran separadamente.

$$ E = E_{vibracional} + E_{rotacional} $$

En consistencia con esta aproximación, la parte vibracional de la energía se describe por el término cuadrático del desarrollo de $V(R)$ alrededor de $R_0$, y la energía de la molécula se aproxima como:

$$ E_{v, l} = -D_e + \hbar \omega_e \left( v + \frac{1}{2} \right) + \frac{\hbar^2 l(l+1)}{2\mu R_0^2} $$

En donde $D_e$ es el mínimo de la energía potencial, $\omega_0$ la frecuencia en la aproximación armónica y $v = 0, 1, 2...$. Notemos que en esta ecuación, se verifica que $E_{rotacional} << E_{vibracional}$, es decir que la contribución centrífuga de la energía cinética solo perturba levemente a la energía electrónica cuando $l$ es suficientemente pequeño.

De acuerdo con la diferencia apreciable entre las escalas de energías, los niveles $E_{v,l}$ presentan características de bandas de niveles vibracionales, ensanchados por la perturbación rotacional.

Esta aproximación pierde exactitud a medida que consideramos estados de mayor momento angular y mayor amplitud de vibración. Las distancias entre los niveles $E_{v+1,l'} - E_{v,l}$ disminuye para valores crecientes de la energía, y los efectos de acople de rotación y vibración se vuelven claramente visibles. 

El procedimiento seguido para estudiar las correcciones a estos niveles, esta basado en el uso de un potencial electrónico analítico, construído fenomenológicamente, con el que se pueda encontrar soluciones de la ecuación de Schrödinger, como el potencial de Morse.

$$ V_M(R) = D_e \left[e^{-2a(R-R_0)} - 2e^{-a(R-R_0)}\right]$$

La ecuación de Schrodinger con el potencial de Morse puede ser resuelta en forma cerrada si se ignora la rotación ($l=0$). Los autovalores de la energía son

$$E_{v,0} = -D_e + \hbar \omega_0(v + \frac{1}{2}) - \frac{1}{4D_e} \left[ \hbar \omega_0 (v + \frac{1}{2}) \right]$$

Al considerar la rotación $l\neq0$, la ecuación de Schrödinger no tiene solución cerrada, por lo que debe usarse un procedimiento aproximado. Se halla el mínimo del potencial efectivo:

$$ U_l =  D_e \left[e^{-2a(R-R_0)} - 2e^{-a(R-R_0)}\right] + \frac{l(l+1)\hbar^2}{2\mu R^2}$$

Y se obtiene asi, la separación internuclear de equilibrio:

$$ R_0^l = R_0 \left( 1 + \frac{B(l)}{a^2 R_0^2 D_e} \right) \qquad B(l) \equiv \frac{l(l+1)h^2}{2\mu R^2} $$

Este valor luego se utiliza para evaluar la profundidad del potencial efectivo:

$$ D_e^l = D_e - B \left[1 - \frac{1}{a^2R_0^2}\left(\frac{B(l)}{D_e} \right) \right]$$

Que se interpreta como el coeficiente de amplitud $D_e^l$ de un nuevo potencial de Morse. Una vez definidas la nueva posición del mínimo y la profundidad del potencial, para cada valor de $l$, se deduce la expresión de energía sustituyendo $D_e \rightarrow D_e^l$, obteniendose:

$$ E_{v, l} = -D_e + \left(v + \frac{1}{2} \right) \hbar \omega_0 \left[1 - \frac{1}{4D_e}\left(v + \frac{1}{2} \right) \hbar \omega_0\right] $$

$$ - \left( \frac{l(l+1)\hbar^2}{4\mu R_0^2 D_e} \right)\left(v + \frac{1}{2} \right) \hbar \omega_0 \left(v + \frac{1}{2} \right) + \frac{l(l+1)\hbar^2}{2\mu R_0^2} \left[1 - \frac{l(l+1)\hbar^2}{\mu^2R_0^2 \omega_0^2} \right] 

Podemos interpretar los términos del resultado como:

- El segundo término contiene la corrección de la dinámica vibracional debido a la anarmónicidad del potencial.

- El tercer término refleja el acople vibro-rotacional.

- El último término contiene una contribución a la energía rotacional pura que puede ser interpretado como un "efecto centrífugo" que produce el estiramiento de la molécula. 

### Espectroscopia



Mucho de lo que se sabe de las moléculas viene de examinar su espectro. Debido a que una molécula es inherentemente más compleja que un átomo, así también lo es la espectroscopia molecular comparada con la espectroscopia atómica. 

Las escalas de energías de las transiciones electrónicas, vibraciones y rotacionales, son distintas, del orden de los $10$, $10^{-1}$ y $10^{-3} \text{eV}$ respectivamente.

#### Transiciones rotacionales

Debido a que la energía rotacional es tan pequeña, su espectro aparece como bandas de líneas muy poco separadas. Esta es la característica más notable de la espectroscopia rotacional.

Un nivel de energía molecular, es según las aproximaciones realizadas anteriormente:

$$ E = E_{0} + \hbar \omega_0 \left( v + \frac{1}{2} \right) + Bj(j+1) $$

En donde:

$$ B = \frac{\hbar^2}{2\mu R_0^2} \qquad \mu = \frac{m_1 m_2}{m_1 + m_2}$$

Es la constante rotacional, con $\mu$ la masa reducida. La primera transición considerada es el cambio estado rotacional puro. Debido a consideraciones de simetría, una transición rotacional pura es un tal que $j$ cambia, pero los estados electrónicos y vibracionales no varían. 

Estas no pueden ocurrir para moléculas homonucleares como el $\text{H}_2$, pero si para moleculas heteronucleares como el $\text{LiF}$ debido a su momento dipolar eléctrico permanente. 

Hay una regla de selección que prohibe transiciónes distintas a:

$$ \Delta j = \pm 1 $$

Para las transiciones con  $j \rightarrow j + 1$, las frecuencias son:

$$ h \nu = B \left[(j+1)(j+2) - j(j+1) \right] = 2B(j + 1) $$

Para las transiciones con $j \rightarrow j - 1$, las frecuencias son:

$$ h \nu = B \left[j(j+1) - j(j-1) \right] = 2Bj $$

En ambos casos, las frecuencias varian con $j$ y la separación de las lineas del espectro es:

$$ \Delta (h\nu) = 2B $$

#### Transiciones vibracionales

Las transiciones vibracionales, no tomar lugar en una molécula homonuclear sin un cambio electrónico asociado, pero pueden en una molécula con un momento dipolar permanente. En este caso las reglas de transición son:

$$ \Delta v = \pm 1 \qquad \Delta j = \pm 1 $$

En una transición vibracional. la absorción siempre implica $\Delta v = +1$, dado a que la energía vibracional es mucho mas grande que la rotacional. En la situacion usuar, el estado inicial es siempre $v=0$, dado a que los estados vibracionales no son poblados a temperatura ambiente. 

Luego, los cambios de energia para las transiciones $j \rightarrow j + 1$ son:

$$ h \nu_R = \hbar \omega_0 \left( 1 + \frac{1}{2} \right) + B(j+1)(j+2) - \frac{1}{2}\hbar \omega_0 - Bj(j+1)$$
$$ = \hbar \omega_0 + 2B(j+1) \qquad j = 1, 2, ... $$

Y  para las transiciones $j \rightarrow j - 1$:

$$ h \nu_P = \hbar \omega_0 \left( 1 + \frac{1}{2} \right) + B(j-1)j - \frac{1}{2}\hbar \omega_0 - Bj(j+1)$$
$$ = \hbar \omega_0 + 2Bj \qquad j = 1, 2, ... $$



En donde los subindices $R$ y $P$ en las frecuencias, son conocidos como las ramas $R$ y $P$. Cada rama forma un grupo de lineas, que estan relacionadas entre sí en terminos de energía o frecuencia.

Las líneas tienen una disposición similar a las líneas producidas por transiciones rotacionales puras. Estas transiciones ocurren cuando una molécula pasa de un estado rotacional a otro.

Las frecuencias de estas transiciones no están en el rango típico de las rotaciones puras (que suelen caer en la región de microondas del espectro). En su lugar, están centradas alrededor de $\hbar \omega_0$, que corresponde a una energía mayor asociada a una frecuencia vibratoria característica y suelen caen en la región del infrarrojo del espectro electromagnético.

La figura muestra como el espectro roto-vibracional divide las líneas del espectro en dos ramas de acuerdo a las ecuaciones anteriores. Notese el gap entre las dos bandas en la frecuencia $\omega_0 2 / \pi$ Este existe puesto a que la transición con $\Delta j = 0$ esta prohibida.



#### Transiciones electronicas

En las transiciones electronicas, las frecuencias podrían ser obtenidas simplemente de la ecuación de las transiciones vibracionales, sin embargo, $\omega_0$ y $B$ no suelen ser las mismas para los estados electrónicos inicial y final. La constante rotacional $B$ depende del momento de inercia molecular, y por ende de la separación internuclear $R_0$, la cual varía para los estados electrónicos. La frecuencia vibraciones $\omega_0$, depende de la curvatura de la energía electrónica en la vecindad de $R_0$, y por ende también varía para los estados electrónicos. Podemos describir entonces la absorción como:

$$ h \nu = E_{s, s'} + \hbar \omega_0' \left( v' + \frac{1}{2} \right) - \hbar \omega_0 \left( v + \frac{1}{2} \right) $$

En donde $\omega_0$ y $\omega_0'$ son las frecuencias vibracionales para los estados electronicos inicial y final respectivamente, $s$ y $v$ son el estado electronico inicial y vibracion  y $s'$ y $v'$ son el estado electronico final y vibracion.

Esta ecuación, configura la escala de las transiciones, las cuales van desde el infra-rojo hasta el ultravioleta. La estructura del espectro es determinada examinando las transiciones rotacionales que ocurren simultaneamente en el cambio $h \nu_{ev}$.

Lo que la estructura rotacional hace es separar las líneas electrónico-vibracional en una banda con bandas con poco espaciamiento. La regla de selección nuevamente es $\Delta j = \pm 1$. 

Por lo que la rama $R$ corresponde a las transiciones $j \rightarrow j + 1$ y cambios de energia:

$$ h \nu_R = h \nu_{ev} + B'(j+1)(j+2) - Bj(j+1) \qquad j = 0, 1, 2, ... $$

La rama $P$ corresponde a las transiciones con $j \rightarrow j - 1$ y cambios de energía:

$$ h \nu_P = h \nu_{ev} + B'j(j-1) - Bj(j+1) \qquad j = 1, 2, 3, ... $$

Debido a que $B \neq B'$, los terminos cuadraticos no se cancelan, y el espectro ya no se ve como una serie de lineas equiespaciadas.

$$ h \nu_R = h \nu_{ev} + (B' - B)j^2 + (3B' - B)j + 2B' $$

$$ h \nu_P = h \nu_{ev} + (B' - B)j^2 - (B' + B)j $$
