### Problema 1

Demostrar que para el caso del ion molecular $\text{H}_2^+$ y al considerar la ecuación de Schrödinger electrónica, como una combinación lineal de los orbitales $1s$ de los átomos de hidrógeno, se puede escribir la integral de Coulomb como:

$$ M = \frac{1}{4\pi\epsilon_0R} \left( 1 - \left(1 + \frac{R}{a_0} \right) e^{-2R/a_0} \right) $$

En donde $a_0$ es el radio de Bohr y $R$ es la distancia entre los núcleos.

### Solución

Se plantea la ecuación de Schrödinger: 

$$ \hat{H} \ket{\psi_{\pm}} = \left( -\frac{\hbar^2}{2m_e} \nabla_r^2 - \frac{ke^2}{|\vec{r} + \vec{R}/2|} - \frac{ke^2}{|\vec{r} - \vec{R}/2|} + \frac{ke^2}{|\vec{R}|} \right) \ket{\psi_{\pm}}$$

$$ \hat{H} \ket{\psi_{\pm}} = (\hat{T} + \hat{V_a} + \hat{V_b} + \hat{V_R}) \ket{\psi_{\pm}}$$

Con las funciones de onda, propuestas como combinacion lineal de los orbitales $1s$ del átomo de hidrógeno:

$$ \ket{\psi_{\pm}} = c_{\pm} \left( \ket{\phi_a} \pm \ket{\phi_b} \right) $$

En donde $\ket{\phi_a}$ y $\ket{\phi_b}$ son tales que:

$$ \langle \vec{r} \ket{\phi_a} = \phi(\vec{r} - \vec{R}/2) \qquad \langle \vec{r} \ket{\phi_b} = \phi(\vec{r} + \vec{R}/2) $$

Las autoenergías del hamiltoniano vendrán dadas por:

$$ E_{\pm} = \bra{\psi_{\pm}} \hat{H} \ket{\psi_{\pm}} = c_{\pm}^2 \bra{\phi_a \pm \phi_b} \hat{T} + \hat{V_a} + \hat{V_b} + \hat{V_R} \ket{\phi_a \pm \phi_b} $$

$$ = c_{\pm}^2 ( \bra{\phi_a} \hat{T} + \hat{V_a} \ket{\phi_a} + \bra{\phi_a} \hat{V_b} + \hat{V_R} \ket{\phi_a} + \bra{\phi_b} \hat{T} + \hat{V_b} \ket{\phi_b} +  \bra{\phi_b} \hat{V_a} + \hat{V_R} \ket{\phi_b} $$

$$ \pm \bra{\phi_a} \hat{T} + \hat{V_b}  \ket{\phi_b} \pm \bra{\phi_a} \hat{V_a} + \hat{V_R} \ket{\phi_b}  \pm \bra{\phi_b} \hat{T} + \hat{V_a}  \ket{\phi_b} \pm \bra{\phi_a} \hat{V_b} + \hat{V_R} \ket{\phi_a} )$$

Se pueden identificar en la energia las integrales de Coulomb $M$:

$$ M = \bra{\phi_a} \hat{V_b} \ket{\phi_a} / e^2 =  \bra{\phi_b} \hat{V_a} \ket{\phi_b} / e^2  = k \int_{R^3} \frac{\phi_{1s}^2(\vec{r} - \vec{R}/2)}{|\vec{r} + \vec{R}/2|} d^3 r $$

En donde $\ket{\phi_{1s}}$ es el orbital $1s$ del átomo de hidrógeno, dado por:

$$ \phi_{1s}(\vec{r}) = \frac{1}{\pi a_0^3} e^{-r/a_0} $$

Usando coordenadas esferoidales proladas $(u,v,\phi)$, con:

$$ u = \frac{r_a + r_b}{R} \qquad v = \frac{r_a - r_b}{R} \qquad dV = \frac{1}{8}R^3(u^2 -v^2) du dv d\phi $$

Con $r_a = |\vec{r} - \vec{R}/2|$ y $r_b = |\vec{r} + \vec{R}/2|$ y limites $0 \leq \phi \leq 2\pi$, $0 \leq u \leq \infty$ y $-1 \leq v \leq 1$.

Reemplazando en la integral de Coulomb.

$$ M = \frac{k}{8}R^3\int_{0}^{2\pi} \int_{0}^{\infty} \int_{-1}^{1} e^{-R(u+v)/a_0}(u^2 - v^2) du dv d\phi $$
