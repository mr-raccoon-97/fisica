En el problema se nos da la energía de absorción para el tono fundamental, es decir la energía de absorción para $v$ de $0$ a $1$, y también la energía de absorción para el primer sobretono, es decir la energía de absorción para $v$ de $0$ a $2$.

El espectro se encuentra en las siguientes regiones:

**La banda $R$**: Esta región se encuentra en la izquierda del espectro y son abarca a las transiciones con más energía. 

**La banda $P$**: Esta región se encuentra en la derecha del espectro y son abarca a las transiciones con menos energía.

En la siguiente figura se muestra el espectro rotacional de la molécula de $\text{CO}$:



Las transiciones correspondientes a la banda $R$  son aquellas en las que hay transiciones rotacionales $j \rightarrow j + 1$ y tiene saltos de energías:

$$ \Delta E_{R}(j) = \hbar \omega_0 + B'(j+1)(j+2) - Bj(j+1) $$


Mientras que las transiciones correspondientes a la banda $P$ son aquellas en las que hay transiciones rotacionales $j \rightarrow j - 1$ y tiene saltos de energías:

$$ \Delta E_{P}(j) = \hbar \omega_0 + Bj(j-1) - B'j(j+1) $$

La transición $Q$ esta prohibida.

Notemos que podemos hallar las constantes rotacionales $B'$ y $B$ como:

$$ \Delta E_{R}(j) - \Delta E_{P}(j) = 4B'(j + \frac{1}{2}) = 4B'j + 2B' $$

$$ \Delta E_{R}(j) - \Delta E_{P}(j+2) = 4B(j + \frac{3}{2}) = 4Bj + 6 B$$

```python
from numpy import array
from matplotlib.pyplot import subplots

table = [
    (0, 2146, None),
    (1, 2150.7, 2139.2),
    (2, 2154.1, 2135.3),
    (3, 2158.0, 2131.4),
    (4, 2161.8, 2127.1),
    (5, 2165.2, 2123.2),
    (6, 2169.0, 2119.4),
    (7, 2172.4, 2115.5),
    (8, 2175.6, 2111.2),
    (9, 2178.2, 2106.8),
    (10, 2183.0, 2103.0)
]

ER = array([row[1] for row in table])
EP = array([row[2] for row in table])

figure, ax = subplots()
ax.plot(range(1, 11), ER[1:], label='E_R')
ax.plot(range(1, 11), EP[1:], label='E_P')
ax.grid()
ax.legend()
ax.set_xlabel('momento angular j')
ax.set_ylabel('energia')
```

```python
from numpy import polyfit

first_energy_difference = array([ER[j] - EP[j] for j in range(1, 11)])
second_energy_difference = array([ER[j] - EP[j+2] for j in range(0, 9)])

slope, intercept = polyfit(range(1, 11), first_energy_difference, 1)
B1 = slope/4
figure, ax = subplots()
ax.scatter(range(1, 11), first_energy_difference)
ax.plot(range(1, 11), slope*array(range(1, 11)) + intercept)
ax.grid()
ax.set_xlabel('momento angular j')
ax.set_ylabel('diferencia de energía')
figure.show()

slope, intercept = polyfit(range(1, 10), second_energy_difference, 1)
B2 = slope/4
figure, ax = subplots()
ax.scatter(range(1, 10), second_energy_difference)
ax.plot(range(1, 10), slope*array(range(1, 10)) + intercept)
ax.grid()
ax.set_xlabel('momento angular j')
ax.set_ylabel('diferencia de energía')
figure.show()

print(f'B1 = {B1}')
print(f'B2 = {B2}')
```

Luego, podemos hallar $B$ y $B'$ haciendo un ajuste lineal a los datos de la tabla. Esto nos da valores de:

$$ \frac{B}{h} = 1.89 \text{cm}^{-1} \qquad \frac{B'}{h} = 1.93 \text{cm}^{-1} $$

Se puede calcular las separaciones internucleares del $\text{CO}$ a partir de las constantes rotacionales para los diferentes niveles de energia vibracionales. Para ello, se puede usar la siguiente ecuación: 

$$B = \frac{\hbar^2}{2\mu R_0^2} \Rightarrow R_0 = \sqrt{\frac{h^2}{8\pi^2\mu B c}} \qquad \mu = \frac{m_1 m_2}{m_1 + m_2}$$

Reemplazando en el caso del $\text{CO}$:
- $m_H = 1,6726 \times 10^{-27} \text{kg}$
- $m_C \approx 12 m_H$
- $m_O \approx 16 m_H$
- $h = 6.626 \times 10^{-34} \text{J s}$
- $B = 1.89 \text{cm}^{-1} = 1.89 \times 10^{2} \text{m}^{-1}$
- $B' = 1.93 \text{cm}^{-1} = 1.93 \times 10^{2} \text{m}^{-1}$
- $c = 3 \times 10^8 \text{m/s}$

```python
from math import sqrt, pi
h = 6.62607015e-34
c = 3e8
mu = lambda m1,m2: m1*m2/(m1+m2)
R0 = lambda B, m1, m2: sqrt(h**2/(8*pi*B*mu(m1, m2)*c))
mH = 1.6737236e-27
mC = 12*mH
mO = 16*mH

B_h = 189
print(R0(B_h*h, mC, mO) * 10**10, ' Amstrongs Para el nivel con v=0')
B_h = 193
print(R0(B_h*h, mC, mO)* 10**10, 'Amstrongs Para el nivel con v=1')
```
