### Cinemática relativista

La teoría especial de la relatividad establece que las leyes de la física son igualmente válidas en todos los sistemas de referencia inerciales, es decir, aquellos donde se cumple la primera ley de Newton (los objetos mantienen su movimiento en línea recta a velocidad constante a menos que actúe una fuerza). Esto implica que no existe un sistema absoluto en reposo, y cualquier sistema en movimiento constante respecto a uno inercial también es inercial.

Imaginemos, entonces, que tenemos dos sistemas de referencia inerciales, $S$ y $S'$, con $S'$ moviéndose a una velocidad uniforme $v$ con respecto a $S$ (y, por lo tanto, $S$ se mueve a una velocidad $-v$ con respecto a $S'$). Podemos organizar nuestras coordenadas de manera que el movimiento se realice a lo largo del eje común $x$ y $x'$ (Figura 3.1), y sincronizar los relojes maestros en el origen de cada sistema para que ambos marquen cero en el instante en que los dos coinciden (es decir, $t=t' =0$ cuando $x=x' =0$). Supongamos ahora que ocurre algún evento en la posición $(x,y,z)$ y tiempo $t$ en $S$.




Escribimos las coordenadas del evento en $S$ como $(ct,x)$, y en $S'$ como $(ct',x')$. La relación entre las coordenadas de un evento en dos sistemas de referencia inerciales se puede describir mediante la transformación $(ct',x') = \Lambda (ct, x)$, o en forma matricial:

$$\begin{bmatrix} ct' \\ x' \end{bmatrix} = \begin{bmatrix} \gamma & \delta \\ \epsilon & \zeta \end{bmatrix} \begin{bmatrix} ct \\ x \end{bmatrix}$$

Supongamos que el evento ocurre en $x=0$, es decir, en el origen de $S$. Entonces, desde el marco de referencia $S'$ que se mueve a una velocidad $v$, el evento ocurre en $x' = -vt'$.

$$\begin{bmatrix} ct' \\ -vt' \end{bmatrix} = \begin{bmatrix} \gamma & \delta \\ \epsilon & \zeta \end{bmatrix} \begin{bmatrix} ct \\ 0 \end{bmatrix} = \begin{bmatrix} \gamma ct \\ \epsilon c t \end{bmatrix}$$

Obtenmos las ecuaciones:

$$ t' = \gamma t \qquad $$

$$ - v t' = - v \gamma t = \epsilon c t \Rightarrow \epsilon = - \frac{v}{c} \gamma \qquad(I)$$

Supongamos ahora que el evento ocurre en $x'=0$, es decir, en el origen de $S'$. Entonces, desde el marco de referencia $S$, el evento ocurre en $x = vt$.

$$ \begin{bmatrix} ct' \\ 0 \end{bmatrix} = \begin{bmatrix} \gamma & \delta \\ \epsilon & \zeta \end{bmatrix} \begin{bmatrix} ct \\ vt \end{bmatrix} = \begin{bmatrix} \gamma ct + \delta vt \\ \epsilon ct + \zeta vt \end{bmatrix}$$

Reemplazando la ecuación $(I)$ en la ecuación de la fila inferior:

$$ 0 = \epsilon ct + \zeta vt = - \frac{v}{c} \gamma c t + \zeta v t$$

$$ \Rightarrow \zeta = \gamma \qquad (II)$$

Supongamos un rayo de luz que se propaga en la dirección positiva del eje $x$ en $S$. Según la teoría de la relatividad, la velocidad de la luz es la misma en todos los sistemas de referencia inerciales, por lo que la velocidad de la luz en $S'$ también es $c$. Por lo tanto:

$$ \begin{bmatrix} ct' \\ ct' \end{bmatrix} = \begin{bmatrix} \gamma & \delta \\ -\frac{v}{c} \gamma & \gamma \end{bmatrix} \begin{bmatrix} ct \\ ct \end{bmatrix} $$

Luego, la transformación es simétrica, y obtenemos que:

$$ \delta = -\frac{v}{c} \gamma  \qquad (III) $$

Finalmente, el hecho de que el sistema $S'$ se mueva a una velocidad $v$ con respecto a $S$ también implica que el sistema $S$ se mueve a una velocidad $-v$ con respecto a $S'$. Por lo tanto, la transformación inversa es:

$$ \begin{bmatrix} ct \\ x \end{bmatrix} = \begin{bmatrix} \gamma & \frac{v}{c} \gamma \\ \frac{v}{c} \gamma & \gamma \end{bmatrix} \begin{bmatrix} ct' \\ x' \end{bmatrix} = \begin{bmatrix} \gamma & \frac{v}{c} \gamma \\ \frac{v}{c} \gamma & \gamma \end{bmatrix} \begin{bmatrix} \gamma & -\frac{v}{c} \gamma \\ -\frac{v}{c} \gamma & \gamma \end{bmatrix} \begin{bmatrix} ct \\ x \end{bmatrix} $$

Lo que nos dice que el producto de las dos matrices es la matriz identidad, es decir:

$$ \begin{bmatrix} \gamma & \frac{v}{c} \gamma \\ \frac{v}{c} \gamma & \gamma \end{bmatrix} \begin{bmatrix} \gamma & -\frac{v}{c} \gamma \\ -\frac{v}{c} \gamma & \gamma \end{bmatrix} = \begin{bmatrix} \gamma^2 - \frac{v^2}{c^2} \gamma^2 & 0 \\ 0& \gamma^2 - \frac{v^2}{c^2} \gamma^2 \end{bmatrix} = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} $$

$$ \Rightarrow \gamma^2 - \frac{v^2}{c^2} \gamma^2 = 1 \Rightarrow \gamma = \frac{1}{\sqrt{1 - \frac{v^2}{c^2}}} \qquad (IV)$$

Por lo que hemos expresado todos los elementos de la transformación $\Lambda$ en términos de la velocidad relativa $v$ entre los sistemas de referencia $S$ y $S'$. Dicha transformación se conoce como transformación de Lorentz. Su forma general, y la de su inversa $\Lambda^{-1}$, es teniendo en cuenta solo la coordenada $x$:

$$ \Lambda = \gamma \begin{bmatrix} 1 & -\beta \\ -\beta & 1 \end{bmatrix} \qquad \Lambda^{-1} = \gamma \begin{bmatrix} 1 & \beta \\ \beta & 1 \end{bmatrix} $$

En donde $\beta = \frac{v}{c}$ y $\gamma = \frac{1}{\sqrt{1 - \beta^2}}$.
