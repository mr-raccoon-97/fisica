### Espacios curvos en la teoría de la relatividad general

La introducción de la relatividad especial hizo que aparente de que la teoría newtoniana de la gravedad no era correcta.

La fuerza gravitacional $\vec{F}_{1,2}$ sobre una masa $m_1$ por una masa $m_2$, en un instante $t$ esta dada por según Newton por:

$$\vec{F}_{1,2} = -G \frac{m_1 m_2}{|\vec{r}_1(t) - \vec{r}_2(t)|} \hat{r}$$

Esto implica la simultaneidad del tiempo. Sin embargo en la relatividad especial la noción de simultaneidad es diferente en cada marco de referencia inercial y la ley de la gravitación universal de Newton puede ser cierta solo en uno de los marcos de referencia.

La masa gravitacional y la masa inercial está bien testeada, con una precisión del orden de $10^{-13}$. 

Esta igualdad implica que el campo gravitacional puede ser eliminado para un objeto en caída libre desde su marco de referencia, también implica que un campo gravitacional puede ser creado mediante la aceleración. No existe un experimento que distinga entre un campo gravitacional y la aceleración uniforme. Ambos son equivalentes.

Aceptar el principio de equivalencia, implica aceptar que la luz cae bajo la influencia de la gravedad aun sin tener masa.

En un espacio vacío, un rayo de luz se mueve en línea recta bajo un marco de referencia inercial. Si el rayo de luz es observado desde un marco de referencia acelerado, transversal a su dirección, este se curva. 

### Geometría

La superficie de una esfera nos brinda un ejemplo sencillo de una geometría
diferente (no euclidiana), en la cual los resultados anteriormente mencionados
de la geometría en el plano no son válidos y deben ser reemplazados por definiciones y teoremas diferentes. 

En la geometía esferica el concepto de línea recta se reemplaza por el de círculo máximo,
el cual se define como la curva que se obtiene al intersecar la esfera con un plano
que pasa por el centro de la misma. Así, mientras que en el plano dos rectas o bien nunca se cortan (paralelas) o se cortan en un único punto, dos rectas arbitrarias (círculos máximos) en la esfera siempre se cortan en dos puntos (diametralmente opuestos).

La distancia entre dos puntos de la esfera se define como la curva de menor longitud sobre la esfera que une ambos, también llamada geodésica.

### Geométrica euclidiana

Supongamos coordenadas cartesianas. Podemos describir el espacio $2$-dimensional mediante distintos sistemas de coordenadas.

Por ejemplo, podemos describir el espacio mediante coordenadas cartesianas $(x, y)$ o polares $(r, \theta)$

En coordenadas cartesianas, la distancia entre dos puntos $dS$ esta dada por:

$$dS^2 = dx^2 + dy^2$$

En coordenadas polares, la distancia entre dos puntos $dS$ esta dada por:

$$dS^2 = dr^2 + r^2 d\theta^2$$



En general, una geometría puede especificarse mediante lo que llamaremos un elemento de línea. Convencionalmente, el elemento de línea se expresa como una relación cuadrática para $dS^2$.

### Geometría no euclidiana de la esfera.

Un ejemplo de geometría no euclidiana es la superficie de una esfera de radio $a$.

La distancia entre dos puntos cercanos en la superficie $(\theta, \phi)$ y $(\theta + d\theta, \phi + d\phi)$. La distancia entre estos dos puntos es:

$$dS^2 = a^2 d\theta^2 + a^2 \sin^2 \theta d\phi^2$$

Para construir un mapa entre la geometría euclidiana y la geometria esferica, introducimos las coordenadas:

$$ x = x(\theta, \phi) \quad y = y(\theta, \phi) $$

Es claro que existe un número infinito de proyecciones. El ejemplo más simple es una proyección lineal:

$$ x = \frac{L_x \phi}{2 \pi} \quad y = \frac{L_y \theta}{\pi} $$

Con $L_x$ y $L_y$ las longitudes del mapa. El resultado, que se conoce como proyección equirectangular.



Es claro que esta proyección deforma las distancias ´
reales, especialmente en la región de los polos. De hecho, toda proyección de la esfera en el plano deforma en alguna medida las distancias. Existen otras
proyecciones más útiles que preservan algunas propiedades de la geometría de la
esfera en el plano, por ejemplo aquella en la que mantiene la linealidad de la longitud y proyecta no linealmente la latitud.

$$ x = \frac{L\phi}{2\pi} \qquad y = y(\theta) $$

La distancia para este cambio de coordenadas se escribe como:

$$ dS^2 = a^2 \left[ \left(\frac{2\pi}{L}\cos\theta(y)\right)dx^2 + \left(\frac{d\theta}{dy}  \right) ^2 dy^2\right]

En resumen los resultados de una geometría puede definirse a través de un elemento de línea, que nos da la distancia entre puntos infinitesimalmente cercanos.

$$ dS^2 = g_{\mu\nu} dx^\mu dx^\nu $$
