### Electrodinámica cuántica

Una forma de derivar la ecuación de Schrödinger es a través de la relación clásica de energía. 

$$ \frac{\vec{p} \cdot \vec{p}}{2\mu} + V = E $$

Y aplicando las descripciones de la mecánica cuántica:

$$ \vec{p} \rightarrow -i\hbar \vec{\nabla} \qquad E\rightarrow i\hbar\frac{\partial}{\partial t}  $$

Obteniendose la ecuación de Schrödinger:

$$ i\hbar\frac{\partial \psi}{\partial t} = -\frac{\hbar^2}{2\mu}\nabla^2\psi + V\psi $$

### Ecuación de Klein-Gordon

La ecuación de Klein-Gordon es una ecuación relativista que describe partículas de espín 0. Se obtiene a partir de la relación de energía relativista de momento y energía:

$$ E^2 - (\vec{p}\cdot\vec{p}) c^2 = m^2 c^4 \Rightarrow  p^{\mu} p_{\mu} - m^2 c^2 = 0$$ 

Por el momento supongamos una particula libre, es decir se deja afuera la relatividad. En $4$-vecotres eto esto es:

$$ p_{\mu} \rightarrow -i\hbar \partial_{\mu} \Rightarrow  -\hbar^2 \partial_{\mu}\partial^{\mu}\psi + m^2 c^2 \psi = 0 $$

Si reemplazamos las derivadas parciales por sus definiciones:

$$ \partial_{0} = \frac{1}{c} \frac{\partial}{\partial t} \qquad \partial_{x} = \frac{\partial}{\partial x} \qquad \partial_{y} = \frac{\partial}{\partial y} \qquad \partial_{z} = \frac{\partial}{\partial z}  $$

$$ \Rightarrow \partial_{\mu}\partial^{\mu} = \frac{1}{c^2} \frac{\partial^2}{\partial t^2} - \nabla^2 $$

Se obtiene la ecuación de Klein-Gordon:

$$ -\frac{1}{c^2} \frac{\partial^2 \psi}{\partial t^2} + \nabla^2\psi = \frac{m^2 c^2}{\hbar^2} \psi $$

Esta ecuación fue descubiera por Schrödinger incluso antes que la ecuación no relativista, pero fue abandonada, cuando fallo en reproducir los niveles de energía del hidrógeno. El problema es que el electrón tiene espín, y la ecuación de Klein-Gordon se aplica a partículas con espín 0. 

La razón de que esta ecuación sea incompatible con la interpretación estadística de Born que establece que $|\psi(\vec{r})|^2$ representa la probabilidad de encontrar la partícula en el punto $\vec{r}$. Esta dificultad proviene de que la ecuación de Klein-Gordon es de segundo orden en el tiempo. Por ello, Dirac buscó una ecuación consistente con la fórmula relativista de energía-momento, pero de primer orden en el tiempo. Para ello se factoriza:

$$(p^{0})^2 - m^2 c^2 = (p^0 + m c)(p^0 - m c)$$

$$ \Rightarrow p^0 + m c = 0 \quad \text{o} \quad p^0 - m c = 0 $$

Cualquiera de las dos condiciones garantiza la relación energía-momento $p^{\mu} p_{\mu} - m^2 c^2 = 0$. Veamos como llegar a esta ecuación a partir de vectores covariantes $p_{\mu}$, proponiendo $\beta^k$ y $\gamma^{\lambda}$ tales que:


$$ (p^{\mu} p _{\mu} - m^2 c^2) = (\beta^k p_k + mc)(\gamma^{\lambda} p_{\lambda} - mc)$$
$$ = \beta^k \gamma^{\lambda} p_k p_{\lambda} - (\beta^k - \gamma^k)p_k - m^2 c^2 $$ 

En donde $\beta^k$ y $\gamma^{\lambda}$ son coeficientes a determinar. Como no se quieren terminos lineales, se deben elegir $\beta^k = \gamma^k$, de modo que:

$$ (p^{\mu} p _{\mu} - m^2 c^2) = \gamma^k \gamma^{\lambda} p_k p_{\lambda}  - m^2 c^2 $$

Finalmente se eligen coeficientes $\gamma^k$ tales que:

$$ p^{\mu} p _{\mu} = \gamma^k \gamma^{\lambda} p_k p_{\lambda} $$

Debemos ahora anular los términos cruzados, es decir que:


$$ \gamma^{\mu} \gamma^{\nu} + \gamma^{\nu} \gamma^{\mu} = 0 \quad \text{para} \quad \mu \neq \nu $$

Viendo el primer término $ p^{\mu} p _{\mu} = (p^0)^2 - (p^1)^2 - (p^2)^2 - (p^3)^2 $, buscamos tambien que:

$$(\gamma^0)^2 = 1 \qquad (\gamma^1)^2 = (\gamma^2)^2 = (\gamma^3)^2 =- 1 $$

Estas ambas condiciones se pueden escribir con el anticonmutador de los $\gamma^{\mu}$, y usando la metrica de Minkowski $\eta^{\mu\nu}$:

$$ \{\gamma^{\mu}, \gamma^{\nu}\} = 2\eta^{\mu\nu} $$

Las soluciones a este problema son las matrices de Dirac, que son matrices de $4\times 4$:

$$ \gamma^0 = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix} \qquad \gamma^i = \begin{pmatrix} 0 & \sigma^i \\ -\sigma^i & 0 \end{pmatrix} $$

En donde $\sigma^i$ son las matrices de Pauli, $1$ la matriz identidad y $0$ la matriz nula. Luego la relación energía-momento se puede escribir como la ecuación de matrices:


$$ (p^{\mu} p _{\mu} - m^2 c^2) = (\gamma^k p_k + mc)(\gamma^{\lambda} p_{\lambda} - mc)$$

$$ \gamma^{\mu} p_{\mu} - m c = 0 $$

Finalmente, se hace la sustitución de $p_{\mu} \rightarrow -i\hbar \partial_{\mu}$, y se obtiene la ecuación de Dirac:

$$ i\hbar \gamma^{\mu} \partial_{\mu} \psi - m c \psi = 0 \qquad \psi = \begin{pmatrix} \psi_1 \\ \psi_2 \\ \psi_3 \\ \psi_4 \end{pmatrix} $$

En donde la ecuación de onda $\psi$ es un vector de 4 componentes, denominado espinor de Dirac.

### La partícula libre

Supongamos que $\psi$ es independiente de la posición, esto es:

$$ \frac{\partial \psi}{\partial x} = \frac{\partial \psi}{\partial y} = \frac{\partial \psi}{\partial z} = 0 $$

Esta ecuación describe un estado estacionario con momento $\vec{p} = 0$, es decir una partícula en reposo. La ecuación de Dirac se reduce a:

$$ i\hbar \gamma^0 \frac{\partial \psi}{\partial t} - m c \psi = 0 $$

$$ \Rightarrow \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix} \frac{\partial \psi}{\partial t} = -i \frac{mc^2}{\hbar} \psi $$

Luego, para un espinor de Dirac $\psi = \begin{pmatrix} \psi_A \\ \psi_B \end{pmatrix} $, se obtiene el sistema de ecuaciones diferenciales:

$$ \frac{\partial \psi_A}{\partial t} = -i \frac{mc^2}{\hbar} \psi_A \qquad \frac{\partial \psi_B}{\partial t} = i \frac{mc^2}{\hbar} \psi_B $$

Las soluciones a estas ecuaciones son:

$$ \psi_A(t) = \psi_A(0) e^{-i \frac{mc^2}{\hbar} t} \qquad \psi_B(t) = \psi_B(0) e^{i \frac{mc^2}{\hbar} t} $$

Notando que $E = mc^2$, se puede identificar el termino de energía en las soluciones. El termino $\psi_A$ describe una partícula con momento $\vec{p} = 0$ como se esperaba, sin embargo el termino $\psi_B$ describe una partícula con energía negativa, estas funciones de onda representan antipartículas.

Luego las soluciones independientes a la ecuación de Dirac son, ignorando la normalización:

$$ \psi^{(1)} = e^{-i(mc^2/\hbar)t} \begin{pmatrix} 1 \\ 0 \\ 0 \\ 0 \end{pmatrix} \quad \psi^{(2)} = e^{-i(mc^2/\hbar)t} \begin{pmatrix} 0 \\ 1 \\ 0 \\ 0 \end{pmatrix} \quad \psi^{(3)} = e^{i(mc^2/\hbar)t} \begin{pmatrix} 0 \\ 0 \\ 1 \\ 0 \end{pmatrix} \quad \psi^{(4)} = e^{i(mc^2/\hbar)t} \begin{pmatrix} 0 \\ 0 \\ 0 \\ 1 \end{pmatrix} $$

Ellas descirben respectivamente un electrón con espín $+\frac{1}{2}$, un electrón con espín $-\frac{1}{2}$, un positrón con espín $+\frac{1}{2}$ y un positrón con espín $-\frac{1}{2}$.

### Ondas planas

Veamos ahora el caso de ondas planas de la forma:

$$ \psi(x) = ae^{-ik\cdot x} u(k) $$

Se busca hallar un $4$-vector $k^{\mu}$ y su asociado bi-espinor $u(k)$ tal que $\psi(x)$ satisface la ecuación de Dirac. Como la dependencia de $x$ esta confinada en el exponente:

$$ \partial_{\mu} \psi = -i k_{\mu} \psi$$

Poniendo esto en la ecuación de dirac:

$$ (\hbar\gamma^{\mu}k_{\mu} - mc) \psi = 0$$

Reemplazando:

$$\gamma^{\mu}k_{\mu} = \gamma^0 k^0 - \vec{\gamma} \cdot \vec{k} = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}  k^0 - \begin{pmatrix} 0 & \vec{\sigma} \\ -\vec{\sigma} & 0 \end{pmatrix} \vec{k} = \begin{pmatrix} k^0 & -\vec{\sigma} \cdot \vec{k} \\ \vec{\sigma} \cdot \vec{k} & -k^0 \end{pmatrix} $$

$$(\hbar \gamma^{\mu}k_{\mu} - mc) u= \begin{pmatrix} k^0\hbar - mc & - \hbar \vec{\sigma} \cdot \vec{k} \\ \hbar \vec{\sigma} \cdot \vec{k} & -k^0\hbar - mc \end{pmatrix} \begin{pmatrix} u_A \\ u_B \end{pmatrix}$$ 

$$ = \begin{pmatrix} (\hbar k^0 - mc)u_A - \hbar \vec{\sigma} \cdot \vec{k} u_B \\ \hbar \vec{\sigma} \cdot \vec{k} u_A - (\hbar k^0 + mc)u_B \end{pmatrix} $$

En donde $u_A$ y $u_B$ son las componentes del bi-espinor $u(k)$, $u_A$ describe la componente de espín $+\frac{1}{2}$ y $u_B$ las dos componentes de arriba y $u_B$ las dos de abajo. Luego:

$$u_A = \frac{1}{k^0 - mc/\hbar}(\vec{\sigma} \cdot \vec{k})u_B \qquad u_B = \frac{1}{k^0 + mc/\hbar}(\vec{\sigma} \cdot \vec{k})u_A $$

$$\Rightarrow u_A = \frac{1}{(k^0)^2 - (mc/\hbar)^2}(\vec{\sigma} \cdot \vec{k})^2 u_A$$

Luego se puede demostrar que:

$$\Rightarrow u_A = \frac{\vec{k}\cdot\vec{k}}{(k^0)^2 - (mc/\hbar)^2} u_A \Rightarrow (k^0)^2 - (mc/\hbar)^2 = \vec{k}\cdot\vec{k}$$

Esto nos dice que $\hbar k^{\mu}$ es un $4$-vector que satisface la relación relativista de energía-momento. Luego, evidentemente:

$$ k^{\mu} = \pm p^{\mu} / \hbar$$

En donde los signos $\pm$ corresponden a partículas y antipartículas. 

Finalmente de la relación entre $u_A$ y $u_B$ podemos encontrar las funciones de onda de las anti-partículas en terminos de una base elegida para las partículas. Esto es:

$$ u_A = \begin{pmatrix} 1 \\ 0 \end{pmatrix} : \qquad u_B = \frac{\vec{\sigma} \cdot \vec{p}}{\hbar k^0 + mc} \begin{pmatrix} 1 \\ 0 \end{pmatrix}
= \frac{c}{E + mc^2} \begin{pmatrix} p_z \\ p_x + i p_y \end{pmatrix} $$

$$ u_A = \begin{pmatrix} 0 \\ 1 \end{pmatrix} : \qquad u_B = \frac{\vec{\sigma} \cdot \vec{p}}{\hbar k^0 + mc} \begin{pmatrix} 0 \\ 1 \end{pmatrix}  = \frac{c}{E + mc^2} \begin{pmatrix} p_x - i p_y \\ p_z \end{pmatrix}$$


Tambien se pueden encontrar las funciones de onda de las partículas en terminos de una base elegida para las anti-partículas. Repitiendo el procedimiento para:


$$ (\hbar\gamma^{\mu}k_{\mu} + mc) \psi = 0$$

Se obtiene la relación entre $u_B$ y $u_A$:

Luego las funciones de onda de las partículas en terminos de una base elegida para las anti-partículas son:

$$ u_B = \frac{1}{k^0 - mc/\hbar}(\vec{\sigma} \cdot \vec{k})u_A $$

$$ u_B = \begin{pmatrix} 1 \\ 0 \end{pmatrix} : \qquad u_A = \frac{\vec{\sigma} \cdot \vec{p}}{\hbar k^0 + mc} \begin{pmatrix} 1 \\ 0 \end{pmatrix}
= \frac{c}{E + mc^2} \begin{pmatrix} p_z \\ p_x + i p_y \end{pmatrix} $$

$$ u_B = \begin{pmatrix} 0 \\ 1 \end{pmatrix} : \qquad u_A = \frac{\vec{\sigma} \cdot \vec{p}}{\hbar k^0 + mc} \begin{pmatrix} 0 \\ 1 \end{pmatrix}  = \frac{c}{E + mc^2} \begin{pmatrix} p_x - i p_y \\ p_z \end{pmatrix}$$


Luego los bi-espinores $u(k)$ son:

$$ u^{(1)} = N \begin{pmatrix} 1 \\ 0 \\
\frac{c p_z}{E + mc^2} \\ \frac{c (p_x + i p_y)}{E + mc^2} \end{pmatrix} \quad u^{(2)} = N \begin{pmatrix} 0 \\ 1 \\
\frac{c (p_x - i p_y)}{E + mc^2} \\ \frac{c p_z}{E + mc^2} \end{pmatrix}$$


$$ v^{(1)} = N \begin{pmatrix}
\frac{c p_z}{E + mc^2} \\ \frac{c (p_x + i p_y)}{E + mc^2} \\ 1 \\ 0 \end{pmatrix} \quad v^{(2)} = N \begin{pmatrix}
\frac{c (p_x - i p_y)}{E + mc^2} \\ \frac{c p_z}{E + mc^2} \\ 0 \\ 1 \end{pmatrix}$$

En donde $N$ es una constante de normalización. Puede demostrarse que:

$$ N = \sqrt{\frac{E + mc^2}{c}} $$


Y las funciones de onda de las partículas y anti-partículas son:

$$ \psi = ae^{-i(p^{\mu}x_{\mu}/\hbar)} u(k) \quad \text{(particulas)} \quad \psi = be^{i(p^{\mu}x_{\mu}/\hbar)} v(k) \quad \text{(anti-particulas)} $$



### Covariantes bilineales

Las componentes de un espinor de Dirac no transforman como un $4$-vector cuando se transforma de un marco de referencia inercial a otro.

Si se tiene un sistema viajando con una velocidad $v$ en la dirección $x$:

$$ \psi \rightarrow \psi' = S \psi $$

Con $S$ la matriz:

$$ S = a_+ + a_- \gamma^0 \gamma^1 = \begin{pmatrix} a_+ & a_- \sigma \\ a_- \sigma & a_+ \end{pmatrix} = \begin{pmatrix}
    a_+ & 0 & 0 & a_- \\ 
    0 & a_+ & a_- & 0 \\
    0 & a_- & a_+ & 0 \\  
    a_- & 0 & 0 & a_+ \\ 
\end{pmatrix} $$

En donde:

$$ a_{\pm} = \pm \sqrt{\frac{1}{2}(\gamma \pm 1)} \qquad \gamma = \sqrt{1 - v^2/c^2}$$

Sería razonable construir una cantidad escalar de un espinor $\psi$ como:

$$ \psi^{\dagger} \psi = \begin{pmatrix} \psi^*_1 & \psi^*_2 & \psi^*_3 & \psi^*_4 \end{pmatrix}  \begin{pmatrix} \psi_1 \\ \psi_2 \\ \psi_3 \\ \psi_4 \end{pmatrix} = |\psi_1|^2 + |\psi_2|^2 + |\psi_3|^2 + |\psi_4|^2 $$

Sin embargo, esto no es invariante cuando se aplica la transformación:

$$ (\psi^{\dagger} \psi)' = (\psi')^{\dagger} \psi' = \psi^{\dagger} S^{\dagger} S \psi \neq \psi^{\dagger}\psi$$

Se introduce entonces el espinor adjunto $\bar{\psi}$:

$$ \bar{\psi} = \psi^{\dagger} \gamma^0 $$

De modo que:

$$ \bar{\psi} \psi = \psi^{\dagger} \gamma^0 \psi = |\psi_1|^2 + |\psi_2|^2 - |\psi_3|^2 - |\psi_4|^2 $$

Es un invariante relativista. Se puede probar que $S^{\dagger} \gamma^0 S = \gamma^0$ y luego:

$$ \bar{\psi}' \psi' = (\psi')^{\dagger} \gamma^0 \psi' = \psi^{\dagger} S^{\dagger} \gamma^0 S \psi = \psi^{\dagger} \gamma^0 \psi = \bar{\psi} \psi $$

### El fotón

En electrodinámica clásica, él los campos eléctrico y magnético producidos por una densidad de carga $\rho$ y una corriente $\vec{J}$ se describen por las ecuaciones de Maxwell:

$$ \vec{\nabla} \cdot \vec{E} = 4\pi \rho \qquad \vec{\nabla} \times \vec{E} + \frac{1}{c}\frac{\partial \vec{B}}{\partial t} = 0 $$

$$ \vec{\nabla} \cdot \vec{B} = 0 \qquad \vec{\nabla} \times \vec{B} - \frac{1}{c}\frac{\partial \vec{E}}{\partial t} = \frac{4\pi}{c} \vec{J} $$

En notación relativista, podemos escribir $\vec{E}$ y $\vec{B}$ como componentes de un tensor $F^{\mu\nu}$:

$$ F^{\mu\nu} = \begin{pmatrix} 0 & -E_x & -E_y & -E_z \\
E_x & 0 & -B_z & B_y \\
E_y & B_z & 0 & -B_x \\
E_z & -B_y & B_x & 0 \end{pmatrix} $$

En donde $\rho$ y $\vec{J}$ se pueden escribir como una cuadrivector $J^{\mu}$:

$$ J^{\mu} = (\rho c, \vec{J}) $$

Luego las ecuaciones de Maxwell se pueden escribir como:

$$ \partial_{\mu} F^{\mu\nu} = \frac{4\pi}{c} J^{\nu} $$

De la antisimetría de $F^{\mu\nu} = -F^{\nu\mu}$, se puede demostrar la ecuación de continuidad.

$$ \partial_{\mu} J^{\mu} = 0 $$

De las ecuaciones de Maxwell se puede deducir que los campos pueden ser descritos por potenciales:

$$ \vec{B} = \vec{\nabla} \times \vec{A} \qquad \vec{E} = -\vec{\nabla} \phi - \frac{1}{c}\frac{\partial \vec{A}}{\partial t} $$

Si definimos los potenciales como un cuadrivector potencial $A^{\mu}$:

$$ A^{\mu} = (\phi/c, \vec{A}) $$

Podemos escribir el tensor de campo electromagnético como:

$$ F^{\mu\nu} = \partial^{\mu} A^{\nu} - \partial^{\nu} A^{\mu} $$

Podemos escribir las ecuaciones de Maxwell en términos de $A^{\mu}$:

$$ \partial_{\mu} \partial^{\mu} A^{\nu} -  \partial^{\nu} (\partial_{\mu} A^{\mu}) = \frac{4\pi}{c} J^{\nu} $$

En la electrodinámica, los campos son entidades físicas y los potenciales son solo construcciones matemáticas. La virtud de los potenciales es que al resolverse, estas tienen en cuenta automáticamente las ecuaciones inhomogéneas de Maxwell. El problema con estas es que las soluciones no son únicas. Por ejemplo si $A_{\mu}$ es una solución, entonces $A_{\mu} + \partial_{\mu} \lambda$ también lo es, en donde $\lambda$ es una función arbitraria de la función y el tiempo.

Este cambio en los potenciales no afecta los campos, y esto es lo que se conoce como invariancia de gauge.

Esta simetría se puede aprovechar tomando $\partial_{\mu} A^{\mu} = 0$, esto se denomina la condición de Lorenz. En este caso, las ecuaciones de Maxwell se reducen a:

$$ \partial^{\mu} \partial_{\mu} A^{\nu} = \frac{4\pi}{c} J^{\nu} $$

No hay una forma limpia de eliminar la ambiguedad residual $A^{\mu}$. 

Para el fotón libre no hay cargas, por lo que este satisface la ecuación:

$$ \partial^{\mu} \partial_{\mu} A^{\nu} = 0 $$

Lo que se reconoce en este contexto como la ecuación de onda de Klein-Gordon. Como en el caso de la ecuación de Dirac, buscamos soluciones con $4$-momentos $p = (E/c, \vec{p})$:

$$ A^{\mu}(x) = a \epsilon^{\mu}(p) e^{-ipx/\hbar} $$

En donde $\epsilon^{\mu}(p)$ es el vector de polarización. Sustituyendo en la ecuación del potencial se obtiene:

$$ p^{\mu} p_{\mu} = 0 \Rightarrow E = |\vec{p}|c $$

Lo que nos dice que el fotón es una partícula sin masa.

Tambien se puede ver que $\epsilon^{\mu}$ tiene cuatro componentes estas no son todas independientes. La condición de Lorenz requiere que:

$$ p^{\mu} \epsilon_{\mu} = 0 $$

En el gauge de Coulomb, se tiene que:

$$ \epsilon^0 = 0 \qquad \vec{p} \cdot \vec{\epsilon} = 0 $$

Esto es, el fotón es un vector polarizado en su dirección de propagación.

### Reglas de Feynman

Anteriormente se encontraron los siguientes espinores para electrones y positrones:

||Electrones|Positrones|
|-|---|---|
| Ondas planas | $\psi(x) = a e^{-i\frac{p \cdot x}{\hbar}} u^{(s)}(p)$ | $\psi(x) = b e^{i\frac{p \cdot x}{\hbar}} v^{(s)}(p)$ |
| Ecuaciones de Dirac | $i\hbar \gamma^{\mu} \partial_{\mu} \psi - mc \psi = 0$ | $i\hbar \gamma^{\mu} \partial_{\mu} \psi + mc \psi = 0$ |
| Ortogonalidad | $\bar{u}^{(1)} u^{(2)} = 0$ | $\bar{v}^{(1)} v^{(2)} = 0$ |
| Normalización | $\bar{u}^{(s)} u^{(s)} = 2mc$ | $\bar{v}^{(s)} v^{(s)} = 2mc$ |
| Completitud | $\sum_s u^{(s)} \bar{u}^{(s)} = \gamma^{\mu} p_{\mu} + mc$ | $\sum_s v^{(s)} \bar{v}^{(s)} = \gamma^{\mu} p_{\mu} - mc$ |

Para los fotones se tiene:

||Fotones|
|-|---|
| Ondas planas | $A_{\mu}(x) = a \epsilon_{\mu}^{(s)} e^{-i\frac{p \cdot x}{\hbar}} \quad s = 1, 2$ |
| Condición de Lorenz | $p^{\mu} \epsilon_{\mu} = 0$ |
| Ortogonalidad | $\epsilon^{(1)*}_{\mu} \cdot \epsilon^{(2)\mu} = 0$ |
| Normalización | $\epsilon^{(s)*}_{\mu} \cdot \epsilon^{(s)\mu} = -1$ |
| Gauge de Coulomb | $\epsilon^{(s)0} = 0 \quad \vec{p} \cdot \vec{\epsilon}^{(s)} = 0$ |
| Completitud | $\sum_s \epsilon^{(s)}_{\mu} \epsilon^{(s)*}_{\nu} = -\delta_{ij} + p_i p_j $|

Para calcular las amplitudes $\mathcal{M}$ de los procesos, se deben tener en cuenta las reglas de Feynman:

1. Para cada línea externa se asocia un momento $p_1, ..., p_n$ y se dibuja una flecha hacia la siguiente línea indicando la dirección positiva, hacia adelante en el tiempo. Para cada linea interna, se asocia un momento $q_1, ..., q_n$ y se dibuja una flecha al costado de la linea indicando la dirección positiva.

2. Lineas externas. Las lineas externas contribuyen como sige:



3. Cada vértice contribuye con un factor $i g_e \gamma^{\mu}$. La constante $g_e$ esta asociada a la carga del electrón como $g_e = e\sqrt{4\pi/\hbar c}$.

4. Cada linea interna contribuye con un propagador como sigue:

$$ \frac{i(\gamma^{\mu q_{\mu} + mc})}{q^2 - m^2 c^2} \quad \text{(electrones y positrones)}$$ 

$$\frac{-ig_{\mu \nu}}{q^2} \quad \text{(fotones)} $$

5. En cada vértice se conserva la energía y el momento. Escribimos en cada uno la función delta de Dirac:

$$(2\pi)^4 \delta^4(p_1 + p_2 + p_3) $$

6. Para cada momento $q$ escribimos un factor $ \frac{d^4q}{(2\pi)^4} $ e integramos.

7. El resultado incluye un factor $(2\pi)^4 \delta^4(p_1 + p_2 + ... - p_n)$. Cancelamos este factor y multiplicamos por $i$. Lo que queda es $\mathcal{M}$.

### Truco de Casimir

En algunos experimentos, los espines del electrón (o positrón) entrante y saliente están especificados, y se dan las polarizaciones de los fotones. Si este es el caso, lo siguiente que se debe hacer es insertar los espinores y vectores de polarización apropiados en la expresión de $\mathcal{M}$ y calcular $∣\mathcal{M}∣^2$, que es la cantidad que realmente necesitamos para determinar las secciones eficaces y las vidas medias.

Sin embargo, con mayor frecuencia no nos interesan los espines. Un experimento típico comienza con un haz de partículas cuyas orientaciones son aleatorias y simplemente cuenta el número de partículas dispersadas en una dirección dada. En este caso, la sección eficaz relevante es el promedio sobre todas las configuraciones iniciales de espín ($s_i$​) y la suma sobre todas las configuraciones finales de espín ($s_f$). En principio, podríamos calcular $∣\mathcal{M}(s_i​→s_f​)∣^2$ para cada combinación posible y luego realizar la suma y el promedio $\langle | \mathcal{M}|^2 \rangle$.

En la practica es mucho mas sencillo calcular $\langle | \mathcal{M}|^2 \rangle$ directamente, sin tener que evaluar las amplitudes individuales.


### Ejemplo: Dispersión de electron-muón

Veamos un ejemplo de como calcular la amplitud de dispersión de un electrón y un muón a través de un fotón. El diagrama de Feynman es el siguiente:



Aplicando las reglas de Feynman

1. Asociamos los momentos $p_1, p_2, p_3, p_4$ a las lineas externas y $q_1 = k$ a la linea interna.

2. En cada vértice hay dos fermiones, el electrón y el muón dispersándose en un fotón interno. Cada línea externa contribuye con un factor $i g_e \gamma^{\mu}$ y cada fermión contribuye con un espinor o un espinor adjunto dependiendo de si este está entrando o saliendo del vértice.

3. Línea interna contribuye con un propagador $-ig_{\mu \nu}/q^2$ dado que se trata de un fotón.

4. En cada vértice se conserva la energía y el momento. Escribimos para el primer vértice $2\pi \delta^4(p_1 - p_3 - k)$ y para el segundo vértice $2\pi \delta^4(p_2 +k - p_4)$.

5. Para el fotón interno escribimos $d^4k/(2\pi)^4$.

6. Multiplicamos todo por $i$ y obtenemos un integrando:

$$ i\left[ \bar{u}^{(s_3)}(p_3)(i g_e \gamma^{\mu})u^{(s_1)}(p_1) \right] (\frac{-i g_{\mu \nu}}{k^2}) \left[ \bar{u}^{(s_4)}(p_4)(ig_e \gamma^{\nu}) u^{(s_2)}(p_2) \right] (2\pi)^4 \delta^4(p_1 - p_3 - k) (2\pi)^4 \delta^4(p_2 +k - p_4) \frac{d^4k}{2\pi^4}$$

Integramos en el espacio de momento y obtenemos la amplitud de dispersión.

$$\mathcal{M} = -\frac{g_e^2}{(p_1 - p_3)^2} \left[ \bar{u}^{(s_3)}(p_3) \gamma^{\mu} u^{(s_1)}(p_1) \right] \left[ \bar{u}^{(s_4)}(p_4) \gamma_{\mu} u^{(s_2)}(p_2) \right] $$

Veamos ahora que sucede con los espinores. Observemos el siguiente diagrama:



Si colocamos la dirección de movimiento sobre el eje $z$, suponiendo que ambas partículas tienen espín up antes de la colisión los espinores se pueden escribir como:

$$ u^{(s_1)} = \begin{pmatrix} \sqrt{(E_e + mc^2)/c} \\ 0 \\ \sqrt{(E_e - mc^2)/c} \\ 0 \end{pmatrix}
= \begin{pmatrix} a_+ \\ 0 \\ a_- \\ 0 \end{pmatrix}
\qquad
u^{(s_2)} = \begin{pmatrix} 0 \\ \sqrt{(E_{\mu} + mc^2)/c} \\ 0 \\ \sqrt{(E_{\mu} - mc^2)/c} \end{pmatrix}
= \begin{pmatrix} 0 \\ b_+ \\ 0 \\ b_- \end{pmatrix}
$$

Como el fotón tiene espín 1, necesariamente las partículas deben cambiar su espín luego de la colisión para conservar el momento angular. Luego los espinores de las partículas después de la colisión son:

$$ 
u^{(s_3)} = \begin{pmatrix} 0 \\ \sqrt{(E_{e} + mc^2)/c} \\ 0 \\ \sqrt{(E_{e} - mc^2)/c} \end{pmatrix} =  \begin{pmatrix} 0 \\ a_+ \\ 0 \\ a_- \end{pmatrix}
\qquad
u^{(s_4)} =  \begin{pmatrix} \sqrt{(E_{\mu} + mc^2)/c} \\ 0 \\ \sqrt{(E_{\mu} - mc^2)/c} \\ 0 \end{pmatrix} = \begin{pmatrix} b_+ \\ 0 \\ b_- \\ 0 \end{pmatrix}
$$

Notemos ahora que:

$$ \bar{u}^{(s_3)} \gamma^{0}u^{(s_1)} = \begin{pmatrix}0 & a_+ & 0 & a_- \end{pmatrix} \gamma^0 \gamma^0 \begin{pmatrix} a_+ \\ 0 \\ a_- \\ 0 \end{pmatrix} = 0 $$

$$ \bar{u}^{(s_3)} \gamma^{\mu}u^{(s_1)} = \begin{pmatrix}0 & a_+ & 0 & a_- \end{pmatrix}
\begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix} \begin{pmatrix} 0 & \sigma^{\mu} \\ -\sigma^{\mu} & 0 \end{pmatrix} \begin{pmatrix} a_+ \\ 0 \\ a_- \\ 0 \end{pmatrix} $$

$$ = \begin{pmatrix}0 & a_+ \end{pmatrix} \sigma^{\mu} \begin{pmatrix} a_- \\ 0 \end{pmatrix} + \begin{pmatrix} 0 & a_- \end{pmatrix} \sigma^{\mu} \begin{pmatrix}  a_+  \\ 0 \end{pmatrix} = 2a_+ a_- \begin{pmatrix} 0 & 1 \end{pmatrix} \begin{pmatrix} \sigma^{\mu}_{11} & \sigma^{\mu}_{11} \\ \sigma^{\mu}_{21} & \sigma^{\mu}_{22} \end{pmatrix} \begin{pmatrix} 1 \\ 0 \end{pmatrix} = 2a_+ a_- \sigma^{\mu}_{21}$$

Analogamente:

$$ \bar{u}^{(s_4)} \gamma^{\mu}u^{(s_2)} = 2b_+ b_- \sigma^{\mu}_{21} $$

Reemplazando en la amplitud de dispersión:

$$\mathcal{M} = -\frac{g_e^2}{(p_1 - p_3)^2} 2a_+ a_- 2b_+ b_- \sigma^{\mu}_{21} \sigma_{\mu 21} = -\frac{8g_e^2}{(p_1 - p_3)^2} (a_+ a_-) (b_+ b_-) $$

Reemplazamos los valores de $a_+, a_-, b_+, b_-$:

$$ a_+ a_- = \sqrt{\frac{E_e + mc^2}{c}} \sqrt{\frac{E_e - mc^2}{c}} = \sqrt{\frac{E_e^2 - m^2c^4}{c^2}} = |\vec{p}_e| \qquad b_+ b_- = |\vec{p}_{\mu}| $$

Por otra parte:

$$ p_1 = \left( \frac{E_e}{c}, \vec{p}_e \right) \qquad p_3 = \left( \frac{E_e}{c}, -\vec{p}_e \right) $$

Luego si $\vec{p}_e = -\vec{p}_{\mu}$:

$$ (p_1 - p_3) = (0, 2\vec{p}_e) \Rightarrow (p_1 - p_3)^2 = -4\vec{p}_e^2 $$

Por lo que la amplitud de dispersión es finalmente:

$$ \mathcal{M} = \frac{8g_e^2}{-4\vec{p}_e^2} \vec{p}_e \vec{p}_{\mu} = -2g_e^2 $$

Por otra parte veamos ahora que sucede si aplicamos el truco de Casimir. En este caso, no nos interesa los espines de las partículas, por lo que debemos calcular el promedio de la amplitud de dispersión. Para ello, debemos calcular el promedio de la amplitud de dispersión al cuadrado.

$$\mathcal{M} = \frac{g_e^4}{(p_1 - p_3)^4} \left[ \bar{u}^{(s_3)}(p_3) \gamma^{\mu} u^{(s_1)}(p_1) \right] \left[ \bar{u}^{(s_4)}(p_4) \gamma_{\mu} u^{(s_2)}(p_2) \right] \left[ \bar{u}^{(s_3)}(p_3) \gamma^{\nu} u^{(s_1)}(p_1) \right] ^*\left[ \bar{u}^{(s_4)}(p_4) \gamma_{\nu} u^{(s_2)}(p_2) \right]^* $$

Luego debemos manejar cantidades de la forma:

$$ G \equiv \left[ \bar{u}(a)\Gamma_1 u(b)\right] \left[ \bar{u}(a)\Gamma_2 u(b) \right]^* $$

Notemos que el conjugado complejo verifica que:

$$ \left[ \bar{u}(a)\Gamma_2 u(b) \right]^* = \left[ u(a)^{\dagger} \gamma^0 \Gamma_2 u(b)\right]^{\dagger} = u(b)^{\dagger} \Gamma_2^{\dagger} \gamma^{0\dagger} u(a) = \bar{u}(b) \bar{\Gamma}_2 u(a) \qquad \bar{\Gamma}_2 = \gamma^0 \Gamma_2^{\dagger} \gamma^0 $$

$$ \Rightarrow G = \left[ \bar{u}(a)\Gamma_1 u(b)\right] \left[ \bar{u}(b)\bar{\Gamma}_2 u(a) \right] $$

Sumando ahora sobre todas las configuraciones de espínor b, usando la relación de completitud y normalización de los espinores:

$$ \sum_b G = \bar{u}(a) \Gamma_1 \left[ \sum_b u^{(s_b)}(p_b) \bar{u}^{(s_b)}(p_b)  \right] \bar{\Gamma}_2 u(a) $$

$$ = \bar{u}(a) \Gamma_1 \left[ \gamma^{\mu} p_{\mu} + m_bc \right] \bar{\Gamma}_2 u(a) = \bar{u}(a) Q u(a) $$

Haciendo lo mismo para el es espínor a, se obtiene la traza:

$$ \sum_a \sum_b G = \text{Tr}(Q(p_a + m_a c)) = \text{Tr}(\Gamma_1 (p_b + m_b c) \bar{\Gamma}_2 (p_a + m_a c)) $$

En el caso de la dispersión electrón-muón, la amplitud de dispersión al cuadrado es:

$$ \langle | \mathcal{M}|^2 \rangle = \frac{g_e^4}{16\vec{p}_e^4} \text{Tr} \left[ \gamma^{\mu}(p_1 + mc) \gamma^{\nu}(p_3 + mc) \right]  \text{Tr} \left[ \gamma_{\mu}(p_2 + Mc) \gamma_{\nu}(p_4 + Mc) \right] $$

Notemos que:

$$ \text{Tr} [\gamma^{\mu}(p_1 + mc) \gamma^{\nu}(p_3 + mc)] = \text{Tr} [\gamma^{\mu} p_1 \gamma^{\nu} p_3] + mc [\text{Tr}(\gamma^{\mu} p_1 \gamma^{\nu}) + \text{Tr}(\gamma^{\mu} \gamma^{\nu} p_3)] + m^2c^2 \text{Tr}(\gamma^{\mu} \gamma^{\nu}) $$

Luego, finalmente:

$$ \text{Tr}(\gamma^{\mu} p_1 \gamma^{\nu} p_3) = (p_1)_{\gamma} (p_3)_{\sigma} \text{Tr}(\gamma^{\mu} \gamma^{\lambda} \gamma^{\nu} \gamma^{\sigma}) = (p_1)_{\gamma} (p_3)_{\sigma} 4(g^{\mu \lambda} g^{\nu \sigma} - g^{\mu \nu} g^{\lambda \sigma} + g^{\mu \sigma} g^{\nu \lambda})$$

$$ = 4 [p_1^{\mu} p_3^{\nu} - g^{\mu \nu} p_1 \cdot p_3 + p_3^{\mu} p_1^{\nu}] $$

$$ \Rightarrow \text{Tr}(\gamma^{\mu} p_1 \gamma^{\nu} p_3) = 4[p_1^{\mu} p_3^{\nu} + p_3^{\mu} p_1^{\nu} + g^{\mu \nu} (m^2c^2 - p_1 \cdot p_3)] $$

Analogamente para el segundo término. Luego:


$$ \langle | \mathcal{M}|^2 \rangle = \frac{4g_e^4}{(p_1 - p_3)^4}[p_1^{\mu} p_3^{\nu} + p_3^{\mu} p_1^{\nu} + g^{\mu \nu} (m^2c^2 - p_1 \cdot p_3)] [p_2^{\mu} p_4^{\nu} + p_4^{\mu} p_2^{\nu} + g^{\mu \nu} (M^2c^2 - p_2 \cdot p_4)] $$

$$ = \frac{8g_e^4}{(p_1 - p_3)^4}[(p_1 \cdot p_2)(p_3 \cdot p_4) + (p_1 \cdot p_4)(p_2 \cdot p_3) - (p_1 \cdot p_3)M^2c^2 - (p_2 \cdot p_4)m^2c^2 + 2m^2 M^2 c^4] $$

De esta forma se puede demostrar que:

$$\frac{d\sigma}{d\Omega} = \left( \frac{\hbar}{8 \pi Mc}\right)^2  \langle | \mathcal{M}|^2 \rangle 
= \left( \frac{\alpha \hbar}{2 \vec{p}^2 \sin(\theta /2)} \right)^2[m^2c^2 + \vec{p}^2 \cos^2(\theta /2)] $$
