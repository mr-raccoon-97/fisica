Antes del descubrimiento del neutrn, muchos pensaban que el ncleo estaba formado por protones y electrones, y que el nmero atmico era igual al exceso de protones. La desintegracin beta pareca confirmar esta idea: despus de todo, los electrones salen disparados, no implica esto que haba electrones en el interior?

Para estimar el momento mnimo de un electrn confinado en un ncleo (radio $10^{-11}$ cm). A partir de la relacin relativista energa-momento: 

$$E^2 - p^2c^2 = m^2c^4$$

Este resultado convenci a algunos de que el electrn del decaimiento beta no poda estar dando vueltas dentro del ncleo, sino que deba producirse en la propia desintegracin.

### Solucin

La relacin de incertidumbre nos dice que:

$$\Delta x \Delta p \geq \frac{\hbar}{2}$$

La incertidumbre asociada a la posicin viene del radio del nucleo, del orden de $1 \text{fm}$, por lo que:

$$ \Delta p = \frac{\hbar}{2 \Delta x} = \frac{\hbar}{2 \times 10^{-15} \text{m}} = 5.274 \times 10^{-20}$$

Usando la relacin relativista de la energa-momento, la energa asociada a este electrn sera:

$$ E = c \sqrt{(p_{min}^2 + m^2c^2)} \approx 98.663 \text{MeV}$$

La energia maxima con la que puede salir despedido un electron proveniente de la desintegracin beta del Tritio es de $18.6 \text{keV}$, lo que indica que el electrn no puede estar confinado en el ncleo, ya que es tres ordenes de magnitud menor.
