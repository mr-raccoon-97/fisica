### Problema 1

Hallar la amplitud $\mathcal{M}$ de la aniquilación de pares $e^+ + e^- \rightarrow \gamma + \gamma$.

### Solución

El primer diagrama de Feynman asociado a este proceso es el siguiente:





Se asignan los espinores $u(p_1), \bar{v}(p_2), \epsilon_{\mu}^*(p_3)$ y $\epsilon_{\nu}^*(p_4)$ a las líneas del diagrama.

A los vértices se les asignan $-ie\gamma^{\mu}$ y $-ie\gamma^{\nu}$.

A la linea interna se le asigna un propagador $\frac{i(\gamma^{\mu}q_{\mu} + m)}{q^2-m^2c2}$ 

A los vértices se les asignan las deltas $(2\pi)^4\delta^4(p_1-q-p_3)$ y $(2\pi)^4\delta^4(p_2+q-p_4)$.

Integramos para $q$ escribiendo $\frac{d^4q}{(2\pi)^4}$.

Cancelamos la función delta como $(2\pi)^4 \delta^4(p_1 + p_2 - p_3 - p_4)$.

$$ \mathcal{M}_1 = (2\pi)^4 i \int [u(p_1) i g_e \gamma^{\mu} \epsilon^*_{\mu}(p_3)]\frac{i(\gamma^{\mu}q_{\mu} + mc)}{q^2 - m^2c^2} [\bar{v}(p_2)  i g_e \gamma^{\nu} \epsilon^*_{\nu}(p_4) ] \delta^4(p_1-q-p_3) \delta^4(p_2+q-p_4) d^4q$$

$$ \mathcal{M}_1 = g_e^2 u(p_1) \not{\epsilon}^*(p_3) \frac{(\not{p}_1 - \not{p}_3) + mc}{(p_1 - p_3)^2 - m^2c^2} \bar{\nu}(p_2) \not{\epsilon^*(p_4)}$$
