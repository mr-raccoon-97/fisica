### El cálculo de Feynman

Existen tres pruebas experimentales para la interacción entre partículas, la dispersión de partículas, desintegraciones y estados ligados. 

La mecánica cuántica, particularmente no relativista está particularmente bien desarrollada para el caso de estados ligados, por contraste, la teoría de relatividad describe mejor las desintegraciones y la dispersión de partículas.

### Sección eficaz de dispersión

Supongamos una partícula incidente con un parámetro de impacto $b$ y que se dispersa con un ángulo $\theta$ como en la siguiente figura.


Si la partícula incide con un parámetro entre $b$ y $b+db$, esta pasara por una sección entre $\theta$ y $\theta+d\theta$, de manera mas general, si esta pasa por una sección $d\sigma$, esta partícula se dispersara en un ángulo solido $d\Omega$.

De la figura anterior, se pueden obtener las siguientes relaciones:

$$ d\sigma = |b db d\phi| \qquad d\Omega = \sin(\theta) d \theta d \phi $$

Luego, se define la sección eficaz de dispersión como:

$$ D(\theta) \equiv \frac{d\Omega}{d\sigma} = |\frac{b}{\sin(\theta)}\frac{db}{d\theta}|$$

La sección cruzada total es la integral de la sección eficaz de dispersión sobre todo el ángulo sólido:

$$ \sigma = \int d\sigma = \int D(\theta) d\Omega$$

### La regla de oro

Para calcular la tasa de desintegración o la sección eficaz se necesita de:

- **La amplitud de dispersión $\mathcal{M}$**: Contiene toda la información dinámica. Se calcula evaluando los diagramas de Feynman relevantes, utilizando las reglas de Feynman correspondientes a la interacción. Representa la probabilidad intrínseca del proceso, considerando las interacciones fundamentales.



- **El espacio de fase disponible**: Es puramente cinemático y depende de las masas, energías y momentos de los participantes. Este refleja que un proceso tiene mayor probabilidad de ocurrir cuanto más "espacio para maniobrar" hay en el estado final. Está relacionado con las configuraciones posibles del sistema en el estado final.

La Regla de Oro de Fermi establece que la tasa de transición de un proceso está dada por el producto del espacio de fase y el cuadrado absoluto de la amplitud.

### Desintegraciones

Supongamos que una partícula en reposo se desintegra en otras partículas $2, 3, 4, ... n$:

$$ 1 \rightarrow 2 + 3 + 4 + ... + n $$

La tasa de desintegración es:

$$ \Gamma = \frac{S}{2\hbar m_1} \int | \mathcal{M} |^2 (2 \pi)^4 \delta^4 (p_1 - p_2 - ... - p_n) \prod_{j=2}^{n} 2 \pi \delta(p_j^2 - m_j^2 c^2) \theta(p_j^0) \frac{d^4 p_j}{(2\pi)^4} $$

En donde $m_i$ es la masa de la $i$-esima partícula $p_i$ es su cuadrimomento. $S$ es un factor estadístico que corrige para el doble conteo cuando hay partículas idénticas en el estado final. Para cada subgrupo de $s$ partículas idénticas, se divide por $s!$.

El diagrama del proceso esta amplitud se calculará evaluando los diagramas de Feynman correspondientes. El resto del cálculo depende del espacio de fase, que describe las configuraciones posibles para las partículas en el estado final.

Para calcular la tasa de transición o la sección eficaz, es necesario integrar sobre todos los posibles momentos cuatridimensionales $p$ de las partículas salientes, sujetas a ciertas restricciones cinemáticas fundamentales:

- **Restricción de masa en las partículas salientes** : Cada partícula debe estar en su "cáscara de masa" (mass shell). Esto significa que su energía y momento están relacionados por la ecuación: $$ p_j^2 = m_j^2 c^2 \qquad E_j^2 - |\vec{p}_j|^2 c^2 = m_j^2 c^4 $$ Esto se garantiza mediante la introducción de la delta de Dirac $\delta(p_j^2 - m_j^2 c^2)$ en la integral de la tasa de desintegración.

- **Energía positiva de las partículas salientes**: Para que el proceso sea físicamente válido, la energía de cada partícula saliente debe ser positiva: $$ E_j > 0 \qquad p_j^0 > 0 $$ 

- **Conservación de energía y momento**: En el proceso, la energía y el momento total se conservan. Esto se expresa mediante otra función delta $ \delta^4 (p_1 - p_2 - ... - p_n)$ en donde $p_1$ es el cuadrimomento de la partícula inicial y $p_2, p_3, ... p_n$ son los cuadrimomentos de las partículas finales.


En algunos casos, los productos finales pesan más que las partículas iniciales. Por ejemplo, una desintegración $Q \rightarrow P + X$, donde la suma de las masas de $P$ y $X$ es mayor que $Q$, no puede ocurrir porque no hay espacio de fase disponible. Esto significa que la tasa de desintegración es cero.

La regla de oro de Fermi se puede interpretar como: **Todos los resultados que sean consistentes con las tres restricciones cinemáticas naturales tienen, a priori, la misma probabilidad de ocurrir**.

Las integrales pueden resolverse teniendo en cuenta que:

$$ \delta(p^2 - m^2 c^2) = \delta[(p^0)^2 - \vec{p}^2 - m^2 c^2] $$

Usando que $\delta(x^2 - a^2) = \frac{1}{2a} [\delta(x-a) + \delta(x+a)]$ se puede escribir:

$$ \theta(p^0) \delta[(p^0)^2 - \vec{p}^2 - m^2 c^2] = \frac{1}{2\sqrt{\vec{p}^2 + m^2 c^2}} \delta(p^0 - \sqrt{\vec{p}^2 + m^2 c^2}) $$

Luego la ecuación se reduce a:

$$ \Gamma = \frac{S}{2\hbar m_1} \int | \mathcal{M} |^2 (2 \pi)^4 \delta^4 (p_1 - p_2 - ... - p_n) \prod_{j=2}^{n} \frac{1}{2\sqrt{\vec{p}_j^2 + m_j^2 c^2}} \frac{d^3 \vec{p_j}}{(2\pi)^3}$$

$$ \text{con} \qquad p_j^0 = \sqrt{\vec{p}_j^2 + m_j^2 c^2} $$

### Dispersión

Supongamos que las partículas $1$ y $2$ colisionan y se producen las partículas $3$, $4$, ... $n$:

$$ 1 + 2 \rightarrow 3 + 4 + ... + n $$

La sección eficaz de dispersión es:

$$ \sigma = \frac{S\hbar^2}{4\sqrt{(\vec{p}_1 \cdot \vec{p}_2)^2 - (m_1 m_2 c^2)^2}} \int | \mathcal{M} |^2 (2 \pi)^4 \delta^4 (p_1 + p_2 - p_3 - ... - p_n) \prod_{j=3}^{n} 2\pi \delta(p_j^2 - m_j^2 c^2) \theta(p_j^0) \frac{d^4 p_j}{(2\pi)^4} $$

En donde $m_i$ es la masa de la $i$-esima partícula $p_i$ es su cuadrimomento. $S$ es un factor estadístico que corrige para el doble conteo cuando hay partículas idénticas en el estado final. Para cada subgrupo de $s$ partículas idénticas, se divide por $s!$.

De nuevo, se puede simplificar la integral de la sección eficaz de dispersión, teniendo en cuenta las propiedades de la delta dirac:

$$ \sigma = \frac{S\hbar^2}{4\sqrt{(\vec{p}_1 \cdot \vec{p}_2)^2 - (m_1 m_2 c^2)^2}} \int | \mathcal{M} |^2 (2 \pi)^4 \delta^4 (p_1 + p_2 - p_3 - ... - p_n) \prod_{j=3}^{n} \frac{1}{2\sqrt{\vec{p}_j^2 + m_j^2 c^2}} \frac{d^3 \vec{p_j}}{(2\pi)^3}$$

$$ \text{con} \qquad p_j^0 = \sqrt{\vec{p}_j^2 + m_j^2 c^2} $$

### Reglas de determinación de Feynman

Para calcular la amplitud de dispersión $\mathcal{M}$, se deben seguir las reglas de Feynman correspondientes a la interacción. Supongamos particulas $A$, $B$ y $C$ con masas $m_A$, $m_B$ y $m_C$ respectivamente. Todas tienen espín $0$ y cada una es su propia antipartícula. Y consideremos un vertice primitivo para estas particulas.



Se supone $m_A > m_B + m_C$ y $m_A > 0$. Hay tres correcciones a primer orden describiendo las desintegraciones $A \rightarrow B + C$.

Primero hallemos el periodo de vida de la partícula $A$, al nivel mas bajo, luego se analizaran varios procesos de desintegración como $A + A \rightarrow B + B$:

Los procesos $A + B \rightarrow A + B$:

El problema es hallar la amplitud $\mathcal{M}$ asociado a un diagrama de Feynman. Para ello se sigue el siguiente procedimiento:

- **Notación**: Se etiquetan los $4$-momentos entrantes y salientes $p_1, ..., p_n$. Se etiquetan los momentos internos $q_1, q_2, ...$ y se coloca una flecha hacia adentro para seguir la dirección positiva de los momentos.

- **Regla de vértice**: Se asigna un factor $-ig$ a cada vértice, en donde $g$ se llama la constante de acoplamiento. Especifica la fuera de interacción entre las partículas en el vértice. 

- **Propagadores**: Para cada línea interna, se escribe un factor $\frac{i}{q_j^2 - m_j^2c^2}$, en donde $q_j$ es el $4$-momento de una linea y $m_j$ es la masa de la partícula que se propaga. Notese que $q_j^2 \neq m_j^2c^2$. Debido a que la partícula virtual no se encuentra en la cáscara de masa.

- **Conservación de momento**: Para cada vértice, se escribe una función delta de Dirac: $$(2\pi)^4 \delta^4(k_1 + k_2 + k_3)$$ En donde los $k$ son los $4$-momentos de las partículas que entran y salen del vértice. Este factor impone la conservación de la energía y el momento en el vértice, dado a que la suma de los $4$-momentos de las partículas que entran es igual a la suma de los $4$-momentos de las partículas que salen.

- **Integración sobre el momento interno**: Se integra sobre todos los momentos internos usando un factor $$\frac{1}{2\pi} $$

- **Cancelar las funciones delta**: Se cancelan las funciones delta de Dirac y se obtiene la amplitud de Feynman, multiplicando todos los factores $(2\pi)^4\delta^4(p_1 + p_2 + ... - p_n)$. El resultado es $\mathcal{M}$.

### Ejemplo: Desintegración de dos partículas

Supongamos que solo hay dos partículas en el estado final de un proceso de desintegración. Tenemos que la tasa de desintegración está dada por:

$$ \Gamma = \frac{S}{32 \pi^2\hbar m_1} \int | \mathcal{M} |^2 \frac{\delta^4 (p_1 - p_2 - p_3)}{\sqrt{(\vec{p}_2^2 + m_2^2 c^2 )(\vec{p}_3^2 + m_3^2 c^2)}} d^3 \vec{p_2} d^3 \vec{p_3}$$

La delta de Dirac $4$-dimensional se puede escribir como:

$$ \delta^4 (p_1 - p_2 - p_3) = \delta(p_1^0 - p_2^0 - p_3^0) \delta^3(\vec{p}_1 - \vec{p}_2 - \vec{p}_3)$$

Si se supone la particula $1$ en reposo, entonces $p_1^0 = m_1 c^2$ y $\vec{p}_1 = 0$. Luego la integral se reduce a:


$$ \Gamma = \frac{S}{32 \pi^2\hbar m_1} \int | \mathcal{M} |^2 \frac{\delta (m_1c - \sqrt{\vec{p}_2^2 + m_2^2c^2} - \sqrt{\vec{p}_3^2 + m_3^2c^2})\delta(\vec{p}_2 + \vec{p}_3)}{\sqrt{(\vec{p}_2^2 + m_2^2 c^2 )(\vec{p}_3^2 + m_3^2 c^2)}} d^3 \vec{p_2} d^3 \vec{p_3}$$

Luego la integral con respecto a $\vec{p}_3$ es trivial, y se obtiene tomando $\vec{p}_3 = -\vec{p}_2$.

$$ \Gamma = \frac{S}{32 \pi^2\hbar m_1} \int | \mathcal{M} |^2 \frac{\delta (m_1c - \sqrt{\vec{p}_2^2 + m_2^2c^2} - \sqrt{\vec{p}_2^2 + m_3^2c^2})}{\sqrt{(\vec{p}_2^2 + m_2^2 c^2 )(\vec{p}_2^2 + m_3^2 c^2)}} d^3 \vec{p_2} $$

Tomando coordenadas esféricas, $\vec{p}_2 \rightarrow (p, \theta, \phi)$


$$ \Gamma = \frac{S}{32 \pi^2\hbar m_1} \int_0^{2\pi} d\phi \int_0^{\pi} \sin \theta d\theta  \int_0^{\infty} | \mathcal{M} |^2 \frac{\delta (m_1c - \sqrt{p^2 + m_2^2c^2} - \sqrt{p^2 + m_3^2c^2})}{\sqrt{(p^2 + m_2^2 c^2 )(p^2 + m_3^2 c^2)}} p^2 dp $$

Haciendo el reemplazo: 

$$ u \equiv \sqrt{p^2 + m_2^2c^2} + \sqrt{p^2 + m_3^2c^2}  $$

$$ \frac{du}{dp} = \frac{p}{\sqrt{p^2 + m_2^2c^2}} + \frac{p}{\sqrt{p^2 + m_3^2c^2}} = \frac{p u }{\sqrt{p^2 + m_2^2c^2}\sqrt{p^2 + m_3^2c^2}}




$$ \Gamma = \frac{S}{8 \pi \hbar m_1} \int_0^{\infty} | \mathcal{M} |^2 \frac{\delta (m_1c - u)}{\sqrt{(p^2 + m_2^2 c^2 )(p^2 + m_3^2 c^2)}} p^2 dp $$

$$ = \frac{S}{8 \pi \hbar m_1} \int_{c(m_2 + m_3)}^{\infty} | \mathcal{M} |^2 \frac{\delta (m_1c - u)}{u} p du $$

Luego si $u = m_1c$ entonces:

$$ m_1c = \sqrt{p_0^2 + m_2^2c^2} + \sqrt{p_0^2 + m_3^2c^2} $$

$$ m_1^2c^2 = p_0^2 + m_2^2c^2 + 2\sqrt{p_0^2 + m_2^2c^2}\sqrt{p_0^2 + m_3^2c^2} + p_0^2+ m_3^2c^2 $$

$$ (m_1^2c^2 - m_2^2c^2 - m_3^2c^2 - 2 p_0^2)^2 = 4(p_0^2 + m_2^2c^2)(p_0^2 + m_3^2c^2) $$

$$ 4 p_0^4 - 4 p_0^2(m_1^2c^2 - m_2^2c^2 - m_3^2c^2) + (m_1^2c^2 - m_2^2c^2 - m_3^2c^2 ) ^2 = 4p_0^4 + 4p_0^2(m_2^2c^2 + m_3^2c^2) + 4 m_2^2 m_3^2c^4 $$

$$  4 p_0^2 m_1^2c^2 = - (m_1^2c^2 - m_2^2c^2 - m_3^2c^2 ) ^2 + 4 m_2^2 m_3^2c^4 $$

$$ p_0 = \frac{c}{2m_1} \sqrt{4 m_2^2 m_3^2 - (m_1^2 - m_2^2 - m_3^2 ) ^2}

Finalmente, reemplazamos para obtener la tasa de desintegración:

$$ \Gamma = \frac{S | \mathcal{M} |^2}{8 \pi \hbar m_1^3}  \sqrt{4 m_2^2 m_3^2 - (m_1^2 - m_2^2 - m_3^2 ) ^2} $$


### Ejemplo: Dispersión de dos partículas

Supongamos que solo hay dos partículas en el estado final de un proceso de dispersión. 

$$ 1 + 2 \rightarrow 3 + 4 $$

Si el marco de referencia se encuentra en el centro de masa, entonces $\vec{p}_1 = -\vec{p}_2$.

$$ \sqrt{(\vec{p}_1 \cdot \vec{p}_2)^2 - (m_1 m_2 c^2)^2} = \frac{(E_1 + E_2)|\vec{p}_1|}{c} $$

Luego la sección eficaz de dispersión está dada por:

$$ \sigma = \frac{S \hbar^2 c}{64 \pi^2(E_1 + E_2)|\vec{p}_1|}
\int | \mathcal{M} |^2 \frac{\delta^4 (p_1 + p_2 - p_3 - p_4)}{\sqrt{(\vec{p}_3^2 + m_3^2 c^2)(\vec{p}_4^2 + m_4^2 c^2)}} d^3 \vec{p_3} d^3 \vec{p_4}$$

Reescribimos la delta de Dirac $4$-dimensional como:

$$ \delta^4 (p_1 + p_2 - p_3 - p_4) = \delta(\frac{E_1 + E_2}{c} - p_3^0 - p_4^0) \delta^3(\vec{p}_3 + \vec{p}_4)$$

Reemplazando en la integral:

$$ \sigma = \left( \frac{\hbar}{8\pi} \right)^2 \frac{Sc}{(E_1 + E_2)|\vec{p}_1|} \int | \mathcal{M} |^2 \frac{\delta(\frac{E_1 + E_2}{c} - \sqrt{\vec{p}_3^2 + m_3^2c^2} - \sqrt{\vec{p}_3^2 + m_4^2c^2})}{\sqrt{(\vec{p}_3^2 + m_3^2 c^2)(\vec{p}_3^2 + m_4^2 c^2)}} d^3 \vec{p_3} d^3p_3$$

En este caso, $\mathcal{M}$ si tiene dependencias con el angulo solido $d\Omega$. Sin embargo, se busca la sección eficaz:

$$ \frac{d\sigma}{d\Omega} = \left( \frac{\hbar}{8\pi} \right)^2 \frac{Sc}{(E_1 + E_2)|\vec{p}_1|} \int_0^{\infty} | \mathcal{M} |^2 \frac{\delta(\frac{E_1 + E_2}{c} - \sqrt{p^2 + m_3^2c^2} - \sqrt{p^2 + m_4^2c^2})}{\sqrt{(p^2 + m_3^2 c^2)(p^2 + m_4^2 c^2)}} p^2 dp$$

Se puede probar finalmente que:

$$ \frac{d\sigma}{d\Omega} = \left( \frac{\hbar}{8\pi} \right)^2 \frac{S}{(E_1 + E_2)} \frac{|\vec{p}_f|}{|\vec{p}_i|}
