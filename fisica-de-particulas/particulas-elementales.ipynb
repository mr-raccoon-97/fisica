### Introducción a la física de particulas.

La física de partículas ha descubierto que la materia está compuesta de pequeñas piezas llamadas partículas elementales. Todas las partículas de un mismo tipo son exactamente iguales. Gracias a esta simplicidad se han podido comprender las interacciones entre estas partículas y las leyes que las gobiernan.

Dado que es imposible estudiar las partículas elementales directamente debido a su tamaño, se utilizan los siguientes métodos, para estudiarlas de manera indirecta:

- Colisiones: Se hacen chocar partículas y se analizan los resultados, como los ángulos de desviación.

- Desintegraciones: Se estudian las partículas que se producen cuando una partícula se descompone espontáneamente.

- Estados Ligados: Se examinan las propiedades de sistemas compuestos por varias partículas unidas.

Algunas características generales del comportamiento de las partículas elementales no dependen de los detalles específicos de las interacciones entre ellas, sino que se derivan directamente de principios fundamentales como la relatividad y la mecánica cuántica. Por ejemplo, en la relatividad, la energía y el momento se conservan siempre, pero la masa en reposo no.

### La mecanica cuántica y la relatividad como base de la física de partículas.

En la mecánica cuántica, un sistema físico se describe por su estado, que puede representarse de diferentes formas. Los procesos físicos, como la dispersión o la desintegración, implican transiciones entre estados cuánticos. Sin embargo, a diferencia de la física clásica, el resultado de un proceso cuántico no está determinado de manera única por las condiciones iniciales. Solo podemos calcular la probabilidad de que ocurra una transición particular.

La combinación de la relatividad y la mecánica cuántica da lugar a conceptos sorprendentes, como la existencia de antipartículas, la explicación del principio de exclusión de Pauli y el teorema CPT. Estos conceptos son fundamentales para la comprensión de la física de partículas y no pueden ser ignorados en una teoría completa.

### El modelo Estándar

El Modelo Estándar es una teoría que describe las interacciones fundamentales entre las partículas elementales, excepto la gravedad. Se basa en dos familias de partículas fundamentales: quarks y leptones. Esta teoría ha sido ampliamente confirmada por experimentos y se considera un gran logro de la física moderna.

### (1930-1956) Antiparticulas

La mecánica cuántica no relativista se completó en un período sorprendentemente breve, entre 1923 y 1926, pero la versión relativista resultó ser un problema mucho más complejo.

Dirac fue el primero en proponer una teoría cuántica relativista para el electrón en 1928. 

La ecuación de Dirac debía describir electrones libres con energía dada por la fórmula relativista: $$E = \sqrt{p^2c^2 + m^2c^4}$$

Donde $p$ es el momento, $m$ es la masa en reposo y $c$ es la velocidad de la luz. Sin embargo, la solución a la ecuación de Dirac admitía una solución correspondiente con energia negativa $$ E = -\sqrt{p^2c^2 + m^2c^4}$$



Esto significaba que, dado que todo sistema tiende a evolucionar hacia estados de menor energía, el electrón debería "escaparse" hacia estados cada vez más negativos, irradiando una cantidad infinita de energía en el proceso.

Dirac postuló que los estados de energía negativa estaban todos ocupados por un "mar" infinito de electrones. Debido a que este mar siempre está presente y es perfectamente uniforme, no ejerce fuerza neta sobre nada. Los estados de energía positiva, por otro lado serian huecos.

En los años 40, Stuckelberg y Feynman proporcionaron una interpretación mucho más sencilla y convincente de los estados de energía negativa, las soluciones de energía negativa se re-expresan como estados de energía positiva de una partícula diferente (el positrón). 

Mientras tanto, resultó que el dualismo en la ecuación de Dirac implicaba que para cada tipo de partícula debía existir una antipartícula correspondiente, con la misma masa pero carga eléctrica opuesta. Entonces, el positrón es el antielectrón. El antiprotón (con carga negativa) fue observado experimentalmente por primera vez en el Bevatron de Berkeley en 1955, y el antineutrón (neutro) fue descubierto en la misma instalación al año siguiente.

Consideramos a las partículas como materia y a las antipartículas como antimateria. La antimateria es idéntica a la materia en todos los aspectos, excepto en la carga eléctrica y otras propiedades que dependen de la carga. Cuando una partícula y su antipartícula se encuentran, se aniquilan mutuamente, produciendo fotones de alta energía.

### (1934-1947) Mesones

Hay un problema conspicuo al que el modelo 'clásico' no se dirige en absoluto: ¿qué mantiene unido al núcleo? Después de todo, los protones, que tienen carga positiva, deberían repelerse violentamente, considerando que están tan juntos. Evidentemente, debe haber alguna otra fuerza, más poderosa que la fuerza de repulsión eléctrica, que une a los protones (y neutrones) entre sí; los físicos de esa época menos imaginativa la llamaron simplemente la fuerza fuerte.

La primera teoría significativa de la fuerza fuerte fue propuesta por Yukawa en 1934. Yukawa asumió que el protón y el neutrón se atraen entre sí por algún tipo de campo. Este campo debería estar debidamente cuantizado. Debido a que caía entre el electrón y el protón, la partícula de Yukawa pasó a conocerse como mesón (que significa 'peso intermedio') y fueron identificadas experimentalmente en 1937.


#### Piones y Muones

Estudios más detallados de las partículas de rayos cósmicos comenzaron a revelar discrepancias preocupantes.

Se demostró que las partículas de rayos cósmicos
interactuaban muy débilmente con los núcleos
atómicos, lo que contradijo la idea de que debían
ser el mesón de Yukawa, el cual proponia interacciones fuertes. En 1947, esta discordancia se resolvió cuando se descubrió que en los rayos cósmicos hay en realidad dos partículas de peso intermedio: el pion $π$ y el muón $μ$.

El verdadero mesón de Yukawa es el pion ($\pi$), este es producido en la parte superior de la atmosfera, pero es desintegrado mucho antes de llegar al suelo, mientras que el muón, al tener una vida media mucho más larga, llega a la superficie terrestre.

### (1930) Neutrinos y Antineutrinos

En el año 1930. Se había planteado un problema en el estudio de la desintegración beta nuclear. En la desintegración beta, un núcleo radiactivo $A$ se transforma en un núcleo ligeramente más ligero B, con la emisión de un electrón:

$$A \rightarrow B + e^-$$

La conservación de la carga requiere que $B$ tenga una unidad más de carga positiva que $A$. (Ahora nos damos cuenta de que el proceso subyacente aquí es la conversión de un neutrón de $A$, en un protón de $B$; (pero en 1930 el neutrón aún no había sido descubierto). 

Así, el núcleo 'hijo' $B$ se encuentra una posición más adelante en la tabla periódica. Hay muchos ejemplos de desintegración beta: el potasio se convierte en calcio ($^{40}K → ^{40}Ca$), el cobre se convierte en zinc ($^{64}Cu → ^{64}Zn$), el tritio se convierte en helio ($^{3}H → ^{3}He$), y así sucesivamente.

Ahora, es característico de las desintegraciones de dos cuerpos ($A \rightarrow B + e^-$) que las energías salientes estén determinadas cinemáticamente, en el marco del centro de masa. Específicamente, si el núcleo 'padre' $A$ está en reposo, de modo que $B$ y $e^-$ salen en direcciones opuestas con momentos iguales y opuestos, entonces la conservación de la energía dicta que la energía del electrón es:

$$E_e = \frac{m_A^2 - m_B^2 + m_e^2}{2m_A} c^2$$

La energía es constante una vez determinadas las masas de las partículas involucradas. 

Sin embargo, cuando los experimentos se realizaron, mostraron que la energía los electrones varía considerablemente, por lo que la ecuación anterior solo determina la energía máxima que puede tener el electrón, por lo que Pauli propuso que en la desintegración beta se emite una partícula adicional, que no interactúa con la materia, y que lleva la energía faltante. 

Esta partícula fue nombrada posteriormente por Fermi como "neutrino" pero por razones que se veran a continuación, actualmente es conocida como "antineutrino".

En la terminología moderna, la desintegración beta se escribe como:

$$n \rightarrow p + e^- + \bar{\nu}_e$$

Es decir, un neutrón se convierte en un protón, un electrón y un antineutrino.

Volviendo a la desintegración del pión $ \pi \rightarrow \mu + {\nu}$, el muón aparece a 90 grados de la dirección original del pión, y este giro abrupto indica que se produjo otra partícula en la desintegración del pión, que no dejó huellas en la emulsión y debe ser eléctricamente neutra, lo que indica que es un neutrino.

También se descubrió que el muón se desintegra en lo que pareciera ser un electrón y dos neutrinos $ \mu \rightarrow e + 2{\nu} $. La forma en deducir que se trataban de dos neutrinos, fue repetir el experimento una y otra vez, midiendo cada vez la energía del electrón. Si siempre resulta ser la misma, se sabe que solo hay dos partículas en el estado final. Pero si varía, entonces debe haber (al menos) tres.

Para 1950, había evidencia teórica convincente de la existencia de los neutrinos, pero aún no había verificación experimental directa. La razón de esto es que los neutrinos interactúan de manera extraordinariamente débil con la materia.
Para probar que estos existen, se utilizó tanque de agua y observaron la reacción de desintegración beta 'inversa':

$$ \bar{\nu} + p^+ \rightarrow n + e^+ $$

En donde un anti-neutrino y un protón se convierten en un neutrón y un positrón. La partícula producida en la desintegración beta ordinaria es un antineutrino, no un neutrino.

Si el neutrino y el antineutrino fueran la misma partícula, la reacción analoga $\nu + n \rightarrow p + e^- $ debería producirse a la misma tasa. Sin embargo, la reacción inversa no se observó, lo que indica que el neutrino y el antineutrino son partículas distintas.

#### Número leptonico

Los números leptónicos son propiedades cuánticas asignadas a las partículas elementales para ayudar a clasificarlas y entender sus interacciones. I

Para establecer qué reacciones son posibles, los físicos asignaron un número leptónico ( $L = +1$ ) a partículas como el electrón, el muón y el neutrino, y ( $L = -1$ ) a sus correspondientes antipartículas: el positrón, el muón positivo y el antineutrino. A las demás partículas se les asigna un número leptónico de cero.


Luego, se propuso una ley de conservación del número leptónico (similar a la ley de conservación de la carga): en cualquier proceso físico, la suma de los números leptónicos antes de la reacción debe ser igual a la suma después de la reacción.

En la desintegración de una partícula, podemos verificar esta conservación. Por ejemplo:

$$ n \rightarrow p + e^- + \bar{\nu}_e $$

Aquí, el neutrón $n$ se descompone en un protón $p$, un electrón $e^-$ y un antineutrino electrónico $\bar{\nu}_e\$. Antes de la desintegración, el número leptónico es cero, y después de la desintegración, también es cero, ya que el electrón tiene $L = +1$ y el antineutrino $L = -1$.

Las desintegraciones de los piones deben ser escritas como:

$$ \pi^- \rightarrow \mu^- + \bar{\nu}_\mu $$

$$ \pi^+ \rightarrow \mu^+ + \nu_\mu $$

En esta desintegración, el número leptónico se conserva, con el muón $\mu^-$ teniendo $L = +1$ y el antineutrino $\bar{\nu}_\mu$ teniendo $L = -1$.

Y la desintegración del muón como:

$$ \mu^- \rightarrow e^- + \bar{\nu}_e + \nu_\mu $$

$$ \mu^+ \rightarrow e^+ + \nu_e + \bar{\nu}_\mu $$

En esta desintegración, el número leptónico se conserva, con el muón $\mu^-$ teniendo $L = +1$ y el antineutrino $\bar{\nu}_\mu$ teniendo $L = -1$.

Diferencia entre neutrino y antineutrino: La principal diferencia entre el neutrino y el antineutrino es su número leptónico. El neutrino tiene $L = +1$ y el antineutrino tiene $L = -1$. Esto ayuda a los físicos a distinguirlos en las reacciones y a aplicar la conservación del número leptónico correctamente.

#### Partículas extrañas (1947-1960)

En 1947, un descubrimiento revolucionario cambió esta percepción: Rochester y Butler observaron una partícula neutral, ahora llamada $K^0$, que se desintegraba en dos piones cargados (uno positivo y otro negativo) dentro de una cámara de niebla. Esta particula fue llamada kaón.

$$K^0 \rightarrow \pi^+ + \pi^-$$

Mas adelante también se observó la desintración de un kaón cargado:

$$K^+ \rightarrow \pi^+ + \pi^+ + \pi^-$$

Los kaones se comportan en algunos aspectos como piones pesados, por lo que la familia de los mesones se amplió para incluirlos. Con el tiempo, se descubrieron muchos más mesones: el $\eta$, el $\phi$, el $\omega$, los $\rho$, y así sucesivamente.



En la imagen se puede observar un $K^+$ entrando por arriba, que se desintegra en $A$ como $K^+ \rightarrow \pi^+ + \pi^+ + \pi^-$.

En 1950 fue descubierta otra particula, esta vez los productos fueron $p^+$ y $\pi^-$. Evidentemente, esta partícula es más pesada que el protón, y fue llamada $\Lambda^0$.   

$$\Lambda^0 \rightarrow p^+ + \pi^-$$

La partícula lambda $\Lambda$ pertenece a la misma familia de partículas que el protón y el neutrón, llamada familia de los bariones.


Aun no se sabía por qué el protón es estable, se propuso la idea de conservar el número bariónico: un valor asignado a las partículas del tipo de los bariones para mantener la estabilidad en las reacciones. 

Así, al protón y al neutrón se les asignó un número bariónico de +1, mientras que sus partículas opuestas (antipartículas) recibieron -1. La conservación de este número asegura que, en cualquier proceso físico, el número total de bariones se mantenga constante.

Así, la desintegración beta del neutrón:

$$n → p^+ + e^- + \bar{nu}$$ 

Está permitida ($A$ = 1 antes y después), y también lo está la reacción en la que se observó por primera vez el antiprotón:

$$p + p = p + p + p + \bar{p}$$

Con ($A$ = 2 en ambos lados). En cambio, la desintegración del protón $p^+ → e^+ + \gamma$ no está permitida, ya que viola la conservación del número bariónico, lo que concuerda con la estabilidad del protón.

A diferencia de los leptones y bariones, no hay conservación de número para los mesones. 

En lenguaje moderno, las partículas extrañas son producidas por la fuerza fuerte (la misma que mantiene unido el núcleo), pero decaen por la fuerza débil (la que explica la desintegración beta y todos los demás procesos de neutrinos). 

### Extrañeza 

A las partículas extrañas se les asigna un número cuántico llamado "extrañeza" que se conserva en las interacciones fuertes, pero no en las débiles. Por ejemplo en la colisión protón-pión, se podrían producir dos partículas extrañas.

$$p + \pi^- \rightarrow \Lambda^0 + K^+$$

$$ \qquad \quad \rightarrow K^0 + \Sigma^0$$

$$ \qquad \quad \rightarrow K^0 + \Lambda^0$$

En este caso la extrañeza se conserva. Las partículas $K$ transportan una extrañeza $S = +1$, mientras que las $\Lambda$ y $\Sigma$ tienen $S = -1$. Las partículas ordinarias como $\pi$, $p$ o $n$ tienen una extrañeza de cero. Sin embargo cuando las partículas extrañas decaen, la extrañeza no se conserva. Por ejemplo:

$$ \Lambda \rightarrow p + \pi^-$$

$$ \Sigma \rightarrow p^+ + \pi^0$$

$$ \quad \rightarrow n + \pi^+$$
