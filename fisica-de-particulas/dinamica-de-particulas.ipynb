### Electrodinámica cuántica (QED)

La electrodinámica cuántica (QED) es la más antigua, la más simple y la más exitosa de las teorías dinámicas; las demás están conscientemente modeladas a partir de ella. 

Todos los fenomenos electromagnéticos pueden ser descritos pueden ser reducibles al siguiente proceso horizontal:



En esta figura el tiempo fluye horizontalmente, hacia la derecha, por lo que este diagrama se interpreta así: una partícula cargada $e$ entra, emite (o absorbe) un fotón $\gamma$ y sale. Para describir procesos más complicados, simplemente combinamos dos o más réplicas de este vértice primitivo. 

**Möller scattering** : Este proceso involucra la interacción entre dos electrones, donde un fotón es intercambiado entre ellos. Este fenómeno corresponde a lo que en la teoría clásica se llamaría la repulsión de Coulomb entre cargas similares. En QED, se describe como la interacción mediada por el intercambio de un fotón. 



**Bhabha scattering**: Este proceso describe la interacción entre un electrón y un positrón. En este caso, la aniquilación de un electrón y un positrón da lugar a la creación de un fotón, que a su vez puede producir un nuevo par electrón-positrón. 



Usando dos vertices, podemos construir los siguientes diagramas, descrbiendo respectivamente, la aniquilación de un par $e^- + e^+ \rightarrow \gamma + \gamma$ y la creación de un par $e^- + \gamma \rightarrow e^- + e^+$ y la dispersión de compton $e^- + \gamma \rightarrow e^- + \gamma$.



La dispersión de Bhabha y Möller están relacionadas por la simetría de cruce, al igual que los tres procesos que se muestran aquí. En términos de diagramas de Feynman, la simetría de cruce corresponde a girar o rotar la figura. Si permitimos más vértices (solo hay que alcanzar en la bolsa y sacar algunos "tinker toys" adicionales), las posibilidades se multiplican rápidamente.

En cada una de estas figuras, dos electrones entran y dos electrones salen. Ellos también describen la repulsión de cargas similares (dispersión de Möller). Las "entrañas" del diagrama son irrelevantes en cuanto al proceso observado. Las líneas internas (aquellas que comienzan y terminan dentro del diagrama) representan partículas que no son observadas, de hecho, que no se pueden observar sin cambiar completamente el proceso. Las llamamos partículas virtuales. Solo las líneas externas (aquellas que entran o salen del diagrama) representan partículas "reales" (observables). Las líneas externas, entonces, te dicen qué proceso físico está ocurriendo: las líneas internas describen el mecanismo involucrado.

Los diagramas de Feynman son puramente simbólicos, no representan trayectorias de partículas. La dimensión horizontal es el tiempo, pero el espacio vertical no corresponde a la separación física. Por ejemplo, en la dispersión de Bhabha, el electrón y el positrón son atraídos, todo lo que el diagrama dice es: “Había un electrón y un positrón, intercambiaron un fotón, y luego volvieron a ser un electrón y un positrón”.


Cada diagrama de Feynman representa un valor que se calcula usando reglas específicas. Para analizar un proceso, se suman las contribuciones de todos los diagramas relevantes. Aunque hay infinitos diagramas, aquellos con más vértices contribuyen menos debido a la constante de estructura fina, por lo que pueden ignorarse. En QED, usualmente solo se necesitan diagramas con hasta cuatro vértices para obtener precisión suficiente.

Las reglas de Feynman imponen la conservación de la energía y el momento en cada vértice de un diagrama, y por lo tanto, en el diagrama completo. Esto implica que el vértice primitivo de la QED (electrodinámica cuántica) por sí solo no representa un proceso físico posible. Podemos dibujar el diagrama, pero al calcularlo, se le asignará un valor de cero. La razón de esto es puramente cinemática: el proceso $e^- \rightarrow e^+ + \gamma$ violaría la conservación de la energía.

### Chromodinámica cuántica (QCD)

En la chromodinámica, el color juega el rol de carga, y el proceso fundamental analógo al $e -> e + \gamma$ es $q -> q + g$, donde $q$ es un quark y $g$ es un gluón.



Como antes, combinamos dos o más de estos "vértices primitivos" para representar procesos más complejos. Por ejemplo, la fuerza entre dos quarks (que es responsable, en primera instancia, de unir quarks para formar hadrones, e indirectamente, mantener unidos los neutrones y protones para formar un núcleo) se describe, en primer orden, mediante el diagrama:




Decimos que la fuerza entre dos quarks es "mediada" por el intercambio de gluones. A este nivel, la cromodinámica es muy similar a la electrodinámica.

Sin embargo, también existen diferencias importantes, especialmente el hecho de que, mientras que hay un solo tipo de carga eléctrica (puede ser positiva o negativa, claro, pero un solo número es suficiente para caracterizar la carga de una partícula), hay tres tipos de color (rojo, verde y azul). En el proceso fundamental $q \rightarrow q + g$, el color del quark (pero no su sabor) puede cambiar. 

Por ejemplo, un quark arriba azul puede convertirse en un quark arriba rojo. Dado que el color (al igual que la carga) siempre se conserva, esto significa que el gluón debe llevar la diferencia: en este caso, una unidad de azul y menos una unidad de rojo.



Dado que los gluones mismos llevan color (a diferencia del fotón, que es eléctricamente neutro), se acoplan directamente con otros gluones y, por lo tanto, además del vértice fundamental quark-gluón, también tenemos vértices primitivos gluón-gluón; de hecho, de dos tipos: vértices de tres gluones y vértices de cuatro gluones.


Una diferencia clave entre la cromodinámica y la electrodinámica es el tamaño de la constante de acoplamiento. En electrodinámica cuántica (QED), cada vértice introduce un factor de aproximadamente $\alpha=1/137$, por lo que solo es necesario considerar diagramas de Feynman con pocos vértices. Sin embargo, en el caso de la fuerza fuerte, la constante de acoplamiento es mayor que 1, lo que ha dificultado el uso de los diagramas de Feynman.

Un logro importante de la cromodinámica cuántica (QCD) fue descubrir que esta constante de acoplamiento no es realmente constante, sino que depende de la distancia entre partículas (es una "constante de acoplamiento en evolución").

A grandes distancias es grande, pero a distancias muy cortas (menores que el tamaño de un protón), se vuelve pequeña, en un fenómeno conocido como "libertad asintótica". Esto permite que los quarks dentro de un protón o un pión interactúen débilmente, y esta característica fue confirmada experimentalmente. Desde el punto de vista teórico, la libertad asintótica permitió rescatar el cálculo de Feynman como herramienta válida en QCD para altos niveles de energía.

Incluso en electrodinámica, el acoplamiento efectivo depende en cierta medida de la distancia al origen. 

Esto puede entenderse cualitativamente de la siguiente manera. Imaginemos primero una carga puntual positiva 
$q$ incrustada en un medio dieléctrico (es decir, una sustancia cuyas moléculas se polarizan en presencia de un campo eléctrico). 

El extremo negativo de cada dipolo molecular será atraído hacia $q$, y el extremo positivo será repelido, como se muestra en la Figura 2.1. Como resultado, la partícula adquiere un "halo" de carga negativa que cancela parcialmente su campo. En presencia del dieléctrico, entonces, la carga efectiva de cualquier partícula se reduce en cierta medida.

En la teoría cuántica de campos (QED), el vacío actúa como un dieléctrico debido a la creación de pares electrón-positrón virtuales, los cuales también crean una "pantalla" alrededor de la carga. Sin embargo, al acercarse a la carga por debajo de la longitud de onda de Compton del electrón $\lambda = \frac{h}{mc}$, donde $\lambda = \frac{mc}{h}$, esta pantalla desaparece, y la carga efectiva aumenta a distancias cortas.

En cromodinámica cuántica (QCD), ocurre algo similar, pero con una diferencia fundamental: además del acoplamiento quark-gluón, también existen vértices gluón-gluón. Esto introduce diagramas adicionales de polarización de gluones, que reducen la fuerza de acoplamiento a distancias cortas, un fenómeno conocido como libertad asintótica. El acoplamiento efectivo disminuye a cortas distancias en QCD, lo que permite que los quarks en un protón o un pion interactúen débilmente a esas escalas.

La constante de acoplamiento efectiva en QCD depende de un parámetro crítico:

$$\alpha = 2f - \frac{11}{3}n$$

Donde $f$ es el número de sabores de quarks y $n$ es el número de colores. En el Modelo Estándar, $f=6$ y $n=3$, por lo que $\alpha = -2$, lo cual implica que el acoplamiento disminuye a distancias cortas. Esto es la base de la libertad asintótica en QCD.
​
Finalmente, a diferencia de la carga eléctrica en QED, los quarks no se observan aislados debido a que están confinados en combinaciones sin color (como mesones y bariones). Se presume que, en QCD, la energía potencial entre quarks aumenta sin límite cuando se los intenta separar, impidiendo su separación completa. Sin embargo, esta teoría aún no ha demostrado formalmente el confinamiento, ya que dicho fenómeno ocurre en el régimen de largo alcance, donde el cálculo de Feynman no es aplicable.

### Interacciones débiles

No existe un nombre específico para "la propiedad que produce fuerzas débiles", en el mismo sentido en que la carga eléctrica genera fuerzas electromagnéticas y el color genera fuerzas fuertes.

Algunos la llaman "carga débil". Cualquiera que sea el término que se use, todos los quarks y todos los leptones la poseen. (Los leptones no tienen color, por lo tanto no participan en las interacciones fuertes; los neutrinos no tienen carga, así que no experimentan fuerzas electromagnéticas; pero todos ellos participan en las interacciones débiles).
