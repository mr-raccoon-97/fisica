### Propiedades y modelos del núcleo
Las primeras investigaciones del núcleo atómico fueron conducidas por Rutherford, el fundador de la física nuclear.

Rutherford describía la radiactividad como la transformación de la materia dentro de un átomo e identificó dos tipos de radiación:

Partículas alfa: son iones de helio con carga positiva.
Partículas beta: son partículas cargadas que se emiten durante la desintegración de un núcleo.

### Partículas nucleares

Experimentos de dispersión de partículas $\alpha$ mostraron que el radio del núcleo es del orden de $10^{-14} \text{m}$, cuatro órdenes de magnitud menor que el radio del átomo.

Los análisis de rayos X establecieron la conexión entre la carga nuclear y el número atómico de Mendeleev. Se reconoció que:

- La masa de un átomo está concentrada en el núcleo, ya que la masa de los electrones es despreciable.
- La masa atómica, representada por el número másico $\text{A}$, es aproximadamente el doble de la carga nuclear, representada por el número de carga $\text{Z}$.

Rutherford propuso la existencia de una nueva partícula masiva sin carga eléctrica, denominada neutrón, que fue descubierta posteriormente por Chadwick.

Desde entonces, la visión aceptada del núcleo ha sido tratar al protón y al neutrón como partículas fundamentales. Ambas son fermiones de spin $1/2$ y tienen masas similares:

$$ M_p = 938.272 \text{MeV/c}^2, \quad M_n = 939.565 \text{MeV/c}^2 $$

La mayoría de las características nucleares de los protones y los neutrones son idénticas y se describen mediante el modelo de nucleones.

Cada nucleón tiene un momento magnético asociado a su spin, el cual es medido experimentalmente y su valor aceptado es:

$$ \mu_p = 2.79 \mu_N, \quad \mu_n = -1.91 \mu_N $$

donde $\mu_N = \frac{e\hbar}{2m_p}$ es el magnetón nuclear.

### Momento magnético.

El magnetón nuclear es una unidad natural para expresar los momentos dipolares magnéticos de partículas pesadas, la relación entre el momento magnético $\mu$ y su spin nuclear $I$ es:

$$ \mu = g\mu_N \frac{I}{\hbar} $$

En el caso de los protones y neutrones, el spin $I$ denota el spin del nucleon, el cual esta asociado al numero cuantico $1/2$, por lo que se pueden escribir, para el proton y el neutron los factores nucleares:

$$ \frac{g_p}{2} = 2.79, \quad \frac{g_n}{2} = -1.91 $$

Estos valores difieren con la teoria de dirac que predice factores:

$$ \frac{g_p}{2} = 1, \quad \frac{g_n}{2} = 0 $$

Lo cual pone de manifiesto que los nucleones son particulas compuestas, y que su estructura interna es compleja, a diferencia de los electrones. La fuerza que une los nucleos es diferente a la fuerza electrostatica, y se conoce como fuerza nuclear.




### Nucleidos

El núcleo atómico existe en una multitud de diferentes especies conocidas como nucleidos. Cada elemento corresponde a un átomo específico con diferente número de neutrones. Los nucleidos se llaman isótopos entre sí si tienen el mismo número atómico.

Esta variedad de nucleidos tiene distintos valores de masa $A$ dado un número atómico $Z$.

Los números $A$ y $Z$ son los números cuánticos que caracterizan a un nucleido, y definimos también el número $N$ como el número de neutrones en un núcleo, de tal forma que: $$A = Z + N$$

La designación de una especie nuclear se hace mediante la notación: $$^{A}_{Z}X_N$$

Donde $X$ es el símbolo del elemento químico.

### Propiedades nucleares

Se pueden clasificar a los núcleos según su estabilidad, dependiendo de si estos son estables o radiactivos. Estos últimos se desintegran natural o artificialmente, emitiendo radiación.

Una medida importante en los procesos radiactivos es la vida media, la cual es el intervalo de tiempo en el cual la mitad de los núcleos de una muestra radiactiva se desintegran.

Una fuente de información sobre propiedades nucleares es la denominada "Chart of the Nuclides", la cual es una tabla que muestra las propiedades de todos los nucleidos $^A_ZX$ en un gráfico del número de protones $Z$ versus el número de neutrones $N$.

Las especies con el mismo número $A$ se denominan isóbaros, y las especies con el mismo número $N$ pero un $Z$ diferente se denominan isótonos. Las especies marcadas con un punto son estables, y las especies no marcadas son radiactivas.

### Abundancias

Un isótopo que existe en la naturaleza se llama isótopo natural. La abundancia de un isótopo se mide en términos de la fracción de la cantidad total de ese elemento que es ese isótopo y está relacionada con la masa de los isótopos naturales como:

$$ \bar{M} = \sum_{i} f_i M_i $$

En donde $\bar{M}$ es la masa promedio del elemento, $f_i$ es la abundancia del isótopo $i$ y $M_i$ es la masa del isótopo $i$.

### Experimento de Rutherford

La primera determinación de la estructura del núcleo fue realizada por Rutherford mediante experimentos de dispersión, disparando partículas alfa contra una lámina delgada de oro y midiendo cómo estas partículas eran desviadas de su trayectoria.

Para estudiar la estructura y las interacciones nucleares, se utilizaron aceleradores de partículas nucleares, que permiten producir haces de partículas con longitudes de onda de De Broglie suficientemente pequeñas para explorar los detalles del núcleo a distancias cortas. Se pueden acelerar protones, deuterones, partículas alfa y otros iones en haces primarios, mientras que los neutrones se pueden extraer en haces secundarios. Todos estos haces se utilizan para bombardear núcleos a diferentes energías.

El experimento de Rutherford, también conocido como el experimento de dispersión de partículas alfa, consistió en disparar partículas $\alpha$ a una delgada lámina de oro. La mayoría de las partículas $\alpha$ pasaban a través de la lámina sin ser desviadas, pero algunas eran desviadas en ángulos grandes.

Rutherford utilizó la fórmula de dispersión, derivada del análisis de las trayectorias de las partículas $\alpha$ al interactuar con los núcleos de los átomos de oro. La fórmula de dispersión es:

$$\frac{d\sigma}{d\Omega} = \left(\frac{1}{4\pi\epsilon_0} \frac{2 Z _1 Z_2e^2}{m v^2}\right)^2 \frac{1}{\sin^4(\theta/2)} \qquad E = \frac{1}{2}mv^2 $$

En donde la sección eficaz diferencial $d\sigma/d\Omega$ es la probabilidad de que una partícula $\alpha$ sea dispersada en un ángulo sólido $d\Omega$ y es proporcional a la intensidad de la radiación dispersada en esa dirección. $Z_1$ y $Z_2$ son las cargas de las partículas $\alpha$ y los núcleos de oro, respectivamente. $e$ es la carga del electrón, $m$ es la masa de la partícula $\alpha$, $v$ es la velocidad de la partícula $\alpha$, $\theta$ es el ángulo de dispersión y $\epsilon_0$ es la permitividad del vacío.

Cuando la partícula $\alpha$ se acerca al núcleo de oro, la interacción entre las partículas $\alpha$ y los núcleos de oro es atractiva y se puede describir mediante la fuerza de Coulomb. La energía cinética de la partícula $\alpha$ es:

$$ E = \frac{1}{2}mv^2 = \frac{1}{4\pi\epsilon_0} \frac{2 Z _1 Z_2e^2}{r} $$

Sin embargo, esta deja de ser válida cuando esta se acerca a una distancia crítica, ya que entran en juegos las interacciones nucleares.

### Scattering de electrones y el radio nuclear

Por otro lado, para estudiar el tamaño y las distribuciones de carga, es más conveniente utilizar electrones como partículas de prueba, ya que sus interacciones están dominadas por la fuerza electromagnética, que es bien conocida. Los electrones interactúan principalmente con los protones del núcleo debido a la fuerza electrostática entre cargas. Estos experimentos tuvieron que esperar hasta que se desarrollaron aceleradores de electrones con energías suficientes.



La dispersión electrones consiste en disparar un haz de electrones contra un blanco de especie $X$ y un espectrómetro para detectar los electrones dispersados. 

Se representan estas observaciones por la sección cruzada diferencial, $\frac{d\sigma}{d\Omega}$, la cual es la probabilidad de que un electrón sea dispersado en un ángulo sólido $d\Omega$.

La distribución de carga nuclear puede ser modelada como:

 $$ \frac{A}{Ze} \rho_p(r) $$

$$ \frac{4}{3}\pi R^3 = \frac{4}{3}\pi R_0^3A $$

# Masa nuclear y energia de enlace.

Las mediciones de Thompson de la relación $\frac{e}{m}$ fueron capaces de detectar la diferencia entre los iones positivos de un elemento dado. Estas indicaciones tempranas de la existencia de isotopos, mostraron que por ejemplo el neón, puede tener dos números másicos diferentes, $A = 20$ y $A = 22$.

Se usó el experimento de Thomson, con un estimado de las abundancias y se obtuvo el promedio de los números atómicos de 20.2, un valor consistente con el peso atómico del neón, de esta forma, se encontró que el peso atómico de cualquier elemento puede ser explicado como el promedio de las abundancias de sus isotopos.

### Energía en reposo y energía ligante

El concepto de la energía en reposo juega un rol mayor en el comportamiento de los núcleos a diferencia de los átomos. Las contribuciones relativistas constituyen una parte significativa de la energía de los núcleos y no pueden ser despreciadas. En vez de ignorarlas, notamos que los procesos atómicos son de baja energía y que los constituyentes de un átomo no pueden transformarse en diferentes especies, ya que las masas en reposo y las energías mantienen sus identidades. Es consistente dejarlas fuera del análisis. El mismo argumento no se sostiene para los núcleos debido a la excitación nuclear y la transformación entre energías comparables.

La energía ligante es una propiedad de cualquier agregado de partículas unidas por interacciones atractivas. La energía en reposo $Mc^2$ es identificada como la energía relativista total de dicho sistema visto desde su centro de masa, junto con la energía cinética y potencial del sistema.

La energía potencial total debe ser atractiva para mantener el núcleo unido con partículas traídas desde el infinito. La energía de los constituyentes separados en reposo consiste solamente en la suma de las energías en reposo de cada partícula, y esta cantidad debe exceder a $Mc^2$. Por lo tanto, la diferencia entre estas energías nos da la energía ligante del sistema.

$$ E_b = \sum_i M_ic^2 - Mc^2 $$

Notemos que, para un átomo neutral con $Z$ electrones:

$$E_b(\text{atomo}) = [M(\text{nucleo}) + Z m_e - M(\text{atomo})] c^2$$

Para un nucleo:

$$E_b(\text{nucleo}) = [Z M_p + N M_n - M(\text{nucleo})] c^2$$

Podemos eliminar la masa nuclear las relaciones anteriores y obtener la energía de enlace por nucleón:

$$ E_b(\text{nucleo}) = [ZM_p + NM_n + Zm_e - M]c^2 $$

La combinación $M_p + m_e$ puede ser puesta igual a la masa del atomo de hidrogeno, luego:

$$ E_b(^AX) = [ZM(^1\text{H}) + NM_n - M(^AX)]c^2 $$

Esta relación expresa la cantidad deseada en términos de las masas encontradas en la espectrometría de masas.

### Energía de separación

La ecuación para la energía ligante puede ser utilizada para introducir otro concepto denominado "la energía de separación de un neutrón".

$$ E_n(^A_ZX) = [M(^{A-1}_ZX) + M_n - M(^A_ZX)]c^2 $$

Esta cantidad es la energía necesaria para separar un neutrón de un núcleo y nos da una idea de la estabilidad del núcleo. Usando la anterior definición de la energía de enlace, podemos escribir:

$$ E_n(^A_ZX) = E_b(^A_ZX) - E_b(^{A-1}_ZX) $$

### La fórmula semiempírica

La energía de enlace define la posición del estado fundamental en un diagrama de niveles de energias nucleares. Para hallar la energía de enlace por nucleón, se puede utilizar la fórmula semiempírica de Weizsäcker. Se consideran cinco términos:

- El término de volumen, que es proporcional al número cuantico $A$.
- El término de superficie, que es proporcional a $A^{2/3}$.
- El término de Coulomb, que es proporcional a $Z(Z-1)/A^{1/3}$.
- El término de asimetría, que es proporcional a $\frac{(A/2-z)^2}{A}$.

La energía de enlace del agregado de estos términos es:

$$ E_b(^A_ZX) = a_1A - a_2A^{2/3} - a_3\frac{Z(Z-1)}{A^{1/3}} - a_4\frac{(A/2-Z)^2}{A} - \epsilon_5$$

Luego, combinando con la ecuación de energia de ligaduras, la fórmula semiempírica de Weizsäcker se puede escribir como:


$$ M(Z,A) = ZM(^1H) + (A-Z)M_n - \left[ a_1 A - a_2 A^{2/3} - a_3 \frac{Z^2}{A^{1/3}} - a_4 \frac{(A/2 - Z)^2}{A} + \epsilon_5  \right] \frac{1}{c^2}$$


En donde la adición de un término de paridad $\epsilon_5$ es necesaria para ajustar la fórmula a los datos experimentales. Los valores de los coeficientes $a_i$ son obtenidos ajustando la fórmula a los datos experimentales. Un ajuste típico de los coeficientes es:

$$ a_1 = 15.76 \text{MeV}, \quad a_2 = 17.81 \text{MeV}, \quad a_3 = 0.7105 \text{MeV}, \quad a_4 = 98.80 \text{MeV} \quad a_5 = 39 \text{MeV} $$

### Modelo de capas

El núcleo es una estructura compleja formada por $A$ cuerpos interactuando entre sí. Podemos escribir su hamiltoniano como:

$$ \hat{H} = \sum_{i=1}^A - \frac{\hbar^2}{2m_i} \nabla_i^2 + V(\vec{R}_1, \vec{r}_2, \ldots, \vec{r}_A) $$

Si se toma en cuenta la interacción entre de Coulomb:

$$ \hat{H} = \sum_{i=1}^A - \frac{\hbar^2}{2m_i} \nabla_i^2 + 

\sum_{j,i\leq j} V_{nuclear}(|\vec{x}_i - \vec{x}_j|) + \sum_{j, i\leq j } \frac{e^2}{|\vec{x}_i - \vec{x}_j|} $$


Desde un punto de vista formal el potencial de interacción contiene termino a $2, 3, ...$ cuerpos, lo cual hace que su resolución sea técnicamente inviable. Debido a esto se utiliza la teoría del campo medio. Es decir se considera que cada nucleón se mueve en un potencial promedio, el cual es la suma de los potenciales de todos los otros nucleones.

Uno de los aspecto bien establecidos del model nuclear de capas es la existencia de núcleos con números mágicos $$(Z, N = 2, 8, 20, 28, 50, 82, 126)$$ cuyas propiedades son:

- Energías de ligadura muy elavadas
- Energías de separación mayores que en los núcleos vecinos
- Momentos cuadrupolares bajos
- Discontinuidades en las energías de las partículas $\alpha$ emitidas.
