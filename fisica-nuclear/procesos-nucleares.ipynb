## Procesos nucleares

La estructura de un núcleo está definida por los estados del sistema cuántico nuclear. El núcleo revela su estructura mediante una variedad de procesos interactivos en los cuales las transformaciones de los estados cuánticos ocurren como un fenómeno observable.

Uno de estos procesos es la emisión de radiación $\gamma$, que ocurre cuando un núcleo en un estado excitado decae a un estado de menor energía. La radiación $\gamma$ es una onda electromagnética de alta frecuencia y corta longitud de onda.

La radiación también puede tomar forma de fragmentos o partículas eyectadas. Estos fenómenos son llamados desintegraciones $\alpha$ y $\beta$. Los procesos $\alpha$, $\beta$ y $\gamma$ representan las tres variedades específicas de radiactividad. Cada emisión está asociada a cierto tipo de inestabilidad nuclear, y cada una es acompañada por una transición de energía en el núcleo.

La radiactividad implica la desintegración espontánea de un núcleo atómico inestable.

Hemos interpretado la radiactividad como la desintegración de un núcleo inestable. Describimos esto ahora en el lenguaje de la física cuántica y trataremos los fenómenos en términos de transiciones entre estados nucleares.

Los decaimientos $\alpha$ y $\beta$ traen cambios definidos en las especies nucleares; sin embargo, la radiación $\gamma$ solo causa un cambio en los estados de un determinado núcleo.

La carga y el número total de nucleones debe conservarse en cada desintegración. Conocemos que una partícula $\alpha$ es un núcleo de $^4He$, y podemos expresar la desintegración $\alpha$ como:

$$ _Z^A X \rightarrow _{Z-2}^{A-4}Y + _2^4He $$

Para las desintegraciónes $\beta^-$ y $\beta^+$, un electrón y un positrón son emitidos, respectivamente. La desintegración $\beta^-$ se expresa como:

$$ _Z^A X \rightarrow _{Z+1}^{A}Y + e^- + \bar{\nu} $$

$$ _Z^A X \rightarrow _{Z-1}^{A}Y + e^+ + \nu $$

Donde $\bar{\nu}$ y $\nu$ se denominan antineutrinos y neutrinos, respectivamente y son partículas neutras con spin $1/2$. Cada una de estas transformaciones procede de un núcleo base $X$ a un núcleo derivado $Y$, y las tres transiciones pueden ser descritas como:

$$ (Z, N) \rightarrow (Z-2, N-2) \quad \text{para desintegración} \quad \alpha $$
$$ (Z, N) \rightarrow (Z+1, N-1) \quad \text{para desintegración} \quad \beta^- $$
$$ (Z, N) \rightarrow (Z-1, N+1) \quad \text{para desintegración} \quad \beta^+ $$




Las propiedades de desintegración incluidas en las fuentes de datos son para cada núcleo:

- La vida media
- El modo de desintegración
- La energía liberada en la desintegración

Estas propiedades proveen una huella observable para cada radionucleido.



### Ley de desintegración exponencial

La inestabilidad es una propiedad de las especies y la vida media es una medida del tiempo de vida de un núcleo. Sin embargo la vida media, no determina el tiempo exacto de desintegración de un núcleo inestable, ya que la desintegración ocurre como un evento discreto y aleatorio. Tal acto de azar es predecible solo en términos de una probabilidad de ocurrencia de una determinada transición en un intervalo de tiempo dado. 

Sean $\gamma$ la cantidad fundamental que describe la probabilidad por unidad de tiempo de la desintegración $X -> Y + r$, donde $r$ denota la radiación emitida. 

Se puede determinar $\gamma$ experimentalmente si se observa la radiación de una fuente conteniendo varios núcleos de especies $X$. Procedemos definiendo $N(t)$ como el número de núcleos de la especie $X$ de la muestro en el tiempo $t$. 

Consideramos un pequeño cambio $dN$ en un intervalo de tiempo $dt$ e identificamos la cantidad positiva $-dN$, como el número de desintegraciones en dicho intervalo. La cantidad $dN/N$ representa la probabilidad de desintegración en en terminos del numero de desintegraciónes en el tiempo $t$, por lo que podemos escribir:

$$ -\frac{dN}{N} = \gamma dt $$

A la probabilidad de desintegración en el intervalo $t$ y $t + dt$. Esto resulta en la "actividad", o tasa de desintegración, proporcional al número de núcleos presentes. La actividad se define como:

$$ -\frac{dN}{dt} = \gamma N $$

Y podemos obtener la cantidad de núcleos en el tiempo $t$ como:

$$ N(t) = N_0 e^{-\gamma t} $$

Esta cantidad es la ley de desintegración exponencial, donde $N_0$ es el número de núcleos en el tiempo $t = 0$. Un núcleo en una muestra, realizara una transición $X \rightarrow Y$, desde un tiempo $t = 0$ en adelante, con una probabilidad $\gamma$ por unidad de tiempo. Obtenemos entonces el tiempo de vida promedio $\langle t \rangle$ como:

$$ \langle t \rangle = \frac{\int_0^{\infty} t \gamma e^{-\gamma t} dt}{\int_0^{\infty} \gamma e^{-\gamma t} dt} = \frac{1}{\gamma} \equiv \tau$$

$$\Rightarrow N(t) = N_0 e^{-t/\tau}$$


La vida media $\tau_{1/2}$ ya fue definido como el tiempo requerido para que la cantidad de núcleos se reduzca a la mitad de su valor inicial. 

$$N(\tau_{1/2}) = N_0 e^{-\tau_{1/2}/\tau} = \frac{N_0}{2}$$

$$\Rightarrow \tau_{1/2} = \tau \ln 2$$

Los resultados para $\tau$ y $\tau_{1/2}$ son interesantes por que conectan cantidades experimentales y teoricas, la vida de una muestra de un nucleo por una parte y por otro lado, la tasa de desintegración para un solo núcleo. 

Podemos medur la actividad de una fuente contando la tasa de particulas cargadas emitidas usando un detector. Cada particula observada, representa una sola desintegración nuclear. Un contador tipico, esta compuesto por una camara de gas al cual se le aplica un campo electrico. La particula es contada, cuando los atomos del gas en la camar son ionizados por el paso de la particula y sus iones son colectados por el campo aplicado. 

La tasa de particulas contadas debe ser igual a la tasa de desintegración de la muestra $-dN/dt$, dejando de lado la eficiencia del detector. La eficiencia del detector depende de dos factores, la fracción total de angulo solido subtendido por el detector  y la probabilidad de una particula de producir una señal en el circuito. La constante de desintegración puede ser determinada facilmente por dicho experimento, usamos el hecho de que la cantidad medida depende del tiempo:

$$-\frac{dN}{dt} = \gamma N= \gamma N_0 e^{-\gamma t} = \left(-\frac{dN}{dt}\right)_0 e^{-\gamma t}$$

$$ \Rightarrow \log \left( - \frac{dN}{dt} \right) = \log \left( - \frac{dN}{dt} \right)_0 - \gamma t  $$

### Desintegraciones alfa

La radiación de partículas alfa dio evidencia de la inestabilidad del núcleo. La emisión de este tipo de partículas está asociada con núcleos pesados inestables y es raramente vista en elementos con números atómicos $Z < 82$.

El fragmento $\alpha$ emitido y su residuo, están unidos por una fuerza nuclear atractiva de corto alcance, y son repelidos por una interacción culombiana repulsiva. 

La energía de la radiación nuclear se absorbe fácilmente en la materia cuando la radiación está en forma de partículas alfa. Las cargas en movimiento pierden su energía cinética en colisiones ionizantes con los átomos en el medio material. Argumentos cinemáticos familiares nos dicen que estas pérdidas de energía por colisión son mayores para las partículas alfa que para la radiación beta, mucho más ligera. Se define un rango en el medio como la distancia en la que la partícula pierde toda su energía cinética. Un absorbente efectivo para partículas alfa podría ser una lámina de aluminio, una hoja de papel o incluso un volumen de aire. El rango disminuye con la densidad del medio absorbente y aumenta con la energía de la partícula. 

Además de la emisión de partículas alfa, los núcleos pesados pueden experimentar otro tipo de desintegración llamado fisión espontánea. En este proceso, el núcleo se divide en dos fragmentos de tamaño similar.

Las velocidades de las partículas $\alpha$ pueden ser medidas con precisión observando las deflexiones de estas en un campo magnético conocido. Un instrumento utilizado para este propósito es el espectrómetro de masas. La curva rango-energía resultante provee una determinación de la energía de la partícula $\alpha$.
