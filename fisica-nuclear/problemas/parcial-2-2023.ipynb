La masa atmica del $^{176}_{71}\text{Lu}$ es de $175,942689 \text{u}$.

- Decae el $^{176}_{71}\text{Lu}$ por emisin $\beta^{-}$ o $\beta^{+}$?

### Solucin

La energa de ligadura de una especie es igual a la diferencia entre la energa en reposo de los constituyentes y la energa en reposo del agregado. Esto es:

$$ E_b(^A_ZX_N) = \left[Z M(^1H) + (A - Z) M_n - M(^A_ZX_N)  \right] c^2 $$

Dado a que la masa del protn y el neutrn son casi iguales, y ambas son muyo mayores a la masa del electrn:

$$ M(^1H) \approx M_p \qquad M_n \approx M_p $$

Luego, en nuestro el caso del $^{176}_{71}\text{Lu}$

$$ E_b(^{176}_{71}\text{Lu}) = \left[71 M(^1H) + (176 - 71) M_n - M(^{176}_{71}\text{Lu})  \right] c^2 \approx (176 M_p - M(^{176}_{71}\text{Lu})) c^2 $$

Reemplazando:

- $c = 299782458 \text{ms}^{-1}$
- $M_p = 1.67262192368 \times 10^{-27} \text{Kg}$
- $M(^{176}_{71}\text{Lu}) = 175,942689 \text{u}$
- $1 \text{u} = 1.66053906660 \times 10^{-27} \text{Kg}$

```python
def uma_to_kg(mass):
    return mass*1.660e-27

def J_to_eV(energy):
    return energy*1.602e-19


c = 3e8#m/s
Mp = 1.6973#Kg
MX = uma_to_kg(175.943)

Eb = lambda A, MX: (A*Mp - MX)*c**2
print(Eb(176, MX), 'J')
print(J_to_eV(Eb(176, MX)), 'eV')
```

Se obtiene que la energa de ligadura ser de $E_b = 4.3 \text{eV}$.

La desintegracin de un isotopo inestable obedece la relacin:

$$\frac{dN}{dt} = -\lambda N \qquad \lambda = \frac{\log(2)}{\tau}$$

En donde $N$ es el numero de atmos del isotopo en funcin del tiempo y $\tau$ es la vida media de este. Resolviendo la ecuacin diferencial:

$$\int_{N_0}^N\frac{dN}{N} = -\lambda\int_0^t dt \Rightarrow \log(\frac{N}{N_0}) = -\frac{\log(2)}{\tau} t \Rightarrow \frac{N}{N_0} = e^{-\frac{\log(2)}{\tau} t}$$

Reemplazando:
- $\tau = 3,7 \times 10^{10}$
- $ t = 4,6 \times 10^{10}$

```python
from math import exp, log

f = lambda t, tau: exp(-log(2)*t/tau)
tau = 3.7e10
t = 4.6e10
print(f(t, tau)*100, '%')
```

Se encuentra que la fraccin de isotopos de $^{176}_{71}\text{Lu}$ es del $42,25 \%$ con respecto a los que haba en el origen de la tierra.
