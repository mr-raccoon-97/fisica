### Problema 1

* Escirbir la configuración nucleónica de manera separada, para protones y neutrones del $^{13}_{6}C$.

* Determinar el espín nuclear del $^{13}_6C$ de acuerdo al modelo de capas.

### Solución

El $^{13}_6C$ tiene $7$ neutrones y $6$ protones. Llenamos ahora las capas de energía.

* Para los protones:
    * La capa $1s_{1/2}$ se llena con $2$ protones
    * La capa $1p_{3/2}$ se llena con $4$ protones 
    
* Para los neutrones:
    * La capa $1s_{1/2}$ se llena con $2$ neutrones
    * La capa $1p_{3/2}$ se llena con $4$ neutrones 
    * La capa $1p_{1/2}$ se llena con $1$ neutrón

Notemos que las capas de los niveles de energía para los protones están llenas, por lo que estos no contribuyen al espín nuclear.

Por otra parte, hay un neutrón libre en la capa $1p_{1/2}$, es decir con $j = 1/2$. Como este es el único nucleón libre, el espín nuclear del $^{13}_6C$ es $I = 1/2$.

### Problema 2

La masa atómica del $^{176}_{71}\text{Lu}$ es de $175,942689 \text{u}$.

- Hallar la energía de ligadura del $^{176}_{71}\text{Lu}$.

- El $^{176}_{71} \text{Lu}$ es un nucleido inestable con una vida media de $3,7 \times 10^{10}$ años. La edad de la tierra se estima en $4,6 \times 10^{10}$ años. Calcular la fracción de isótopos de $^{176}_{71}$ que se encuentran actualmente en la tierra.

- Decae el $^{176}_{71}\text{Lu}$ por emisión $\beta^{-}$ o $\beta^{+}$?

### Solución

La energía de ligadura de una especie es igual a la diferencia entre la energía en reposo de los constituyentes y la energía en reposo del agregado. Esto es:

$$ E_b(^A_ZX_N) = \left[Z M(^1H) + (A - Z) M_n - M(^A_ZX_N)  \right] c^2 $$

Dado a que la masa del protón y el neutrón son casi iguales, y ambas son muyo mayores a la masa del electrón:

$$ M(^1H) \approx M_p \qquad M_n \approx M_p $$

Luego, en nuestro el caso del $^{176}_{71}\text{Lu}$

$$ E_b(^{176}_{71}\text{Lu}) = \left[71 M(^1H) + (176 - 71) M_n - M(^{176}_{71}\text{Lu})  \right] c^2 \approx (176 M_p - M(^{176}_{71}\text{Lu})) c^2 $$

Reemplazando:

- $c = 299782458 \text{ms}^{-1}$
- $M_p = 1.67262192368 \times 10^{-27} \text{Kg}$
- $M(^{176}_{71}\text{Lu}) = 175,942689 \text{u}$
- $1 \text{u} = 1.66053906660 \times 10^{-27} \text{Kg}$

```python
def uma_to_kg(mass):
    return mass*1.660e-27

def J_to_eV(energy):
    return energy*1.602e-19


c = 3e8#m/s
Mp = 1.6973#Kg
MX = uma_to_kg(175.943)

Eb = lambda A, MX: (A*Mp - MX)*c**2
print(Eb(176, MX), 'J')
print(J_to_eV(Eb(176, MX)), 'eV')
```

Se obtiene que la energía de ligadura será de $E_b = 4.3 \text{eV}$.

La desintegración de un isotopo inestable obedece la relación:

$$\frac{dN}{dt} = -\lambda N \qquad \lambda = \frac{\log(2)}{\tau}$$

En donde $N$ es el numero de atómos del isotopo en función del tiempo y $\tau$ es la vida media de este. Resolviendo la ecuación diferencial:

$$\int_{N_0}^N\frac{dN}{N} = -\lambda\int_0^t dt \Rightarrow \log(\frac{N}{N_0}) = -\frac{\log(2)}{\tau} t \Rightarrow \frac{N}{N_0} = e^{-\frac{\log(2)}{\tau} t}$$

Reemplazando:
- $\tau = 3,7 \times 10^{10}$
- $ t = 4,6 \times 10^{10}$

```python
from math import exp, log

f = lambda t, tau: exp(-log(2)*t/tau)
tau = 3.7e10
t = 4.6e10
print(f(t, tau)*100, '%')
```

Se encuentra que la fracción de isotopos de $^{176}_{71}\text{Lu}$ es del $42,25 \%$ con respecto a los que había en el origen de la tierra.
