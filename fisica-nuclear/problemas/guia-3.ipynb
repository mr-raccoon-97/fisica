### Problema 1

La figura muestra datos experimentales de dispersión de particulas $\alpha$ de $27\text{MeV}$ por $^{208}Pb$.



 A partir de la información de la figura, estime el radio nuclear del $^{208}Pb$.


### Solución


Cuando la partícula $\alpha$ se acerca al núcleo de plomo, la interacción entre la partícula $\alpha$ y el núcleo de plomo deja de ser culombiana, es decir, la energía potencial de la partícula $\alpha$ no puede ser descrita por la formula:

$$ E = \frac{1}{2}mv^2 = k \frac{ Z _1 Z_2e^2}{r} $$

Donde $Z_1$ y $Z_2$ son las cargas de las partículas $\alpha$ y el núcleo, respectivamente, $r$ es la distancia entre las partículas $\alpha$ y el núcleo, $m$ es la masa de la partícula $\alpha$, $v$ es la velocidad de la partícula $\alpha$, $e$ es la carga del electrón y $\epsilon_0$ es la permitividad del vacío.

Si suponemos que la formula deja de ser valida para una distancia critica $R$ igual al radio nuclear, entonces podemos calcular el valor de $R$ a partir de la energia:

$$ R =  k \frac{ Z _1 Z_2 e^2}{E} $$

Reemplazando:

- $E_c \approx 27 \text{MeV}$
- $Z_1 = 2$, el numero atomico de una particula $\alpha$
- $Z_2 = 82$, el numero atomico del plomo.
- $k = 8.9875517873681764 \times 10^9 \, \text{J m}/\text{C}^2 $
- $e = 1.602 \times 10^{-19} \text{C}$.

```python
E = 27e6 #eV
Z1 = 2
Z2 = 82
k = 8.9875e9 #J m / C^2
e = 1.602e-19 #C

def eV_to_J(energy):
    return energy*1.609e-19

R = lambda Z1, Z2, E: k*Z1*Z2*e**2/E

print(R(Z1,Z2,eV_to_J(E))*10**15, 'fm')
```

Obteniéndose un radio nuclear aproximado para el $^{208}\text{Pb}$ de $8.7 \text{fm}$

### Problema 2

En la figura se muestran mediciones de la intensidad dispersada de electrones de $360\text{MeV}$ y $420\text{MeV}$ por un núcleos de $^{12}C$ y $^{16}O$.



Considerando un modelo de difracción de la luz por un objeto circular opaco, estime el radio nuclear de los núcleos de 
$^{12}C$ y $^{16}O$.

### Solución

Si se considera a los electrones como ondas de Broglie, entonces se puede modelar la dispersión de los electrones por los núcleos como un problema de difracción por un objeto circular opaco, en donde el objeto circular opaco es el núcleo. La intensidad de la onda dispersada por el objeto circular opaco es:

$$ \sin(\theta) = 1.22 \frac{\lambda}{D} $$

Donde $\lambda$ es la longitud de onda de la luz y $D$ es el diámetro de la apertura circular, en este caso el diametro del nucleo. La longitud de onda de Broglie se puede hallar de la relación:

$$ E = h \nu = \frac{h c}{\lambda} $$

En la figura, se ven los ángulos críticos en los cuales los electrones fueron dispersados. Podemos hallar el diámetro del núcleo entonces de la relación:

$$ D = 1.22 \frac{h c}{ E \sin(\theta)} $$

Reemplazando los valores criticos:

- $E_{^{16}O} = 420 \text{MeV} \qquad \theta_{^{16}O} \approx 44 \degree$
- $E_{^{12}C} = 420 \text{MeV} \qquad  \theta_{^{12}C} \approx 51 \degree$
- $E_{^{16}O} = 360 \text{MeV} \qquad  \theta_{^{16}O} \approx 53 \degree$


```python
from math import sin, pi
h =  6.626e-34
c = 3e8

def degree_to_radian(angle):
    return pi*angle/180
    
def eV_to_J(energy):
    return energy*1.609e-19

D = lambda E, theta: 1.22*h*c/(E*sin(theta))

for E, theta in [(420, 44), (420, 51), (360, 53)]:
    print(D(eV_to_J(E*10**6), degree_to_radian(theta))*10**15,'fm')
```

Obteniéndose un radio de $5.2\text{fm}$ para el oxígeno y $4.6\text{fm}$ para el carbono.

### Problema 3

Calcular la energia de ligadura del $^4\text{He}$ y del $^{208}\text{Pb}$. Evaluar la energía de ligadura por nucleón.

### Solución

La energía de ligadura de un átomo $^A_ZX$ es igual a la diferencia entre la energía de los nucleones individuales y la energía en reposo del atomo. Esto es:

$$ E_b(^A_ZX) = \left[ Z M(^1H) + NM_n - M(A^X) \right] c^2 $$

En donde $M(^1H) \approx m_p$ es la masa del átomo de hidrógeno aproximadamente igual a la masa del protón, $M_n$ es la masa del neutrón, $M(A^X)$ es la masa del átomo $^A_ZX$ y $N = A - Z$ es el número de neutrones.

Reemplazando los valores:

- $M(^4He) = 4.00260325413 \text{u}$
- $M(^{208}Pb) = 207.9766521 \text{u}$
- $M_n = 1.00866491588 \text{u}$
- $M(^1H) = 1.007276466621 \text{u}$
- $c = 3 \times 10^8 \text{m/s}$

```python
def uma_to_kg(mass):
    return mass*1.660e-27

def J_to_MeV(energy):
    return energy/1.609e-13

MH = 1.6726e-27 #kg
Mn = 1.6749e-27 #kg
c = 299792458 

E = lambda Z, A, MA: (Z*MH + (A-Z)*Mn - MA)*c**2

for Z, A, MA, atom in [(2, 4, 4.0026, 'He'), (82, 208, 207.9766, 'Pb')]:
    print(f'Energía de ligadura para el {atom}',J_to_MeV(E(Z, A, uma_to_kg(MA))), 'MeV')
```

La energía de ligadura para cada átomo es:
- $E_b(^4He) = 28.3 \text{MeV}$
- $E_b(^{208}Pb) = 1647.5 \text{MeV}$

Finalmente la energía de ligadura por nucleón es:

- $E_b(^4He) = 7.1 \text{MeV}$
- $E_b(^{208}Pb) = 7.9 \text{MeV}$

### Problema 4

Calcular la energía de separación de un neutrón para $^{113}Cd$ y $^{114}Cd$. Comparar y analizar los resultados. Las masas atómicas de los isótopos de Cadmio 112, 113 y 114 son respectivamente:

- $M(^{112}Cd) = 111.9027639 \text{u}$
- $M(^{113}Cd) = 112.90440811 \text{u}$
- $M(^{114}Cd) = 113.903365 \text{u}$

### Solución

La energía de separación de un neutrón, será la diferencia entre la energía del isótopo en reposo $^A_ZX_N$ y la energía de los constituyentes, es decir, la energía del isotopo $^{A-1}_ZX_{N-1}$, y la energía en reposo del neutrón. Esto es:

$$ E = \left[ M(^{A-1}_ZX) + M_n - M(^A_ZX) \right] c^2$$

```python
M112Cd = 111.9027639
M113Cd = 112.90440811
M114Cd = 113.903365
Mn = 1.00866491588
c = 3e8

def uma_to_kg(mass):
    return mass*1.660e-27

def J_to_MeV(energy):
    return energy/1.609e-13

E = lambda Mi, Mf: c**2*uma_to_kg(Mf + Mn - Mi) 

print('113Cd -> 112Cd + n: ', J_to_MeV(E(M113Cd, M112Cd)), 'MeV')
print('114Cd -> 113Cd + n: ',J_to_MeV(E(M114Cd, M113Cd)), 'MeV')
```

Se necesitan $6.5\text{MeV}$ y $9.01\text{MeV}$ respectivamente para separar un neutrón de los isotopos $^{113}\text{Cd}$ y $^{114}\text{Cd}$ respectivamente. Esto quiere decir que se necesita casi un $50\%$ más de energía para separar un neutrón del isotopo $^{114}\text{Cd}$, por lo que este será más estable.

### Problema 7

A partir de la fórmula semiempírica de la masa atómica, determinar el nucleido de menor masa atómica para el conjunto de isóbaros de número másico $A$. Grafiar la masa atómica para el conjunto de isóbaros de número másico $A$. Graficar la masa atómica en función del número atómico para los isóbaros de número másico $98$, $99$ y $100$. Analizar la posibilidad de que exista un isótopo estable del $\text{Tc}$ en estos casos.

### Solución

La formula semiempírica de la masa atómica es:

$$ M(Z,A) = ZM(^1H) + (A-Z)M_n - \left[ a_1 A - a_2 A^{2/3} - a_3 \frac{Z^2}{A^{1/3}} - a_4 \frac{(A/2 - Z)^2}{A} + \epsilon_5  \right] \frac{1}{c^2}$$

En donde los valores de los coeficientes son:

$$ \epsilon_5 = \begin{cases} \pm \frac{a_5}{A^{3/4}} & \text{Si } N, Z \text{ son ambos pares o impares} \\ 0 & \text{De lo contrario.} \end{cases} $$

$$ a_1 = 15.76 \text{MeV}, \quad a_2 = 17.81 \text{MeV}, \quad a_3 = 0.7105 \text{MeV}, \quad a_4 = 98.80 \text{MeV} \quad a_5 = 39 \text{MeV} $$

Si se deriva la fórmula, con respecto a $Z$, obtenemos la tasa de cambio de la masa con respecto al número atómico dado un número másico $A$ fijo. Por lo que es posible hallar un minimo igualando la derivada a 0.

$$ \frac{\partial M(Z, A)}{\partial Z} = M(^1H) - M_n - \left[ - a_3\frac{2 Z}{A^{1/3}} + a_4\frac{2 (A/2 - Z)}{A} \right] \frac{1}{c^2} = 0$$

$$ c^2( M(^1H) - M_n ) = \left[ - a_3\frac{2 Z}{A^{1/3}} + a_4 - a_4\frac{2Z}{A} \right] $$

$$ \Rightarrow Z = \frac{A}{2} \frac{a_4 - c^2( M(^1H) - M_n )}{ a_3 A^{2/3} + a_4 } $$

Los isobaros estables con números másicos $98$, $99$ y $100$ son:
- $^{98}_{42}\text{Mo}$ , $^{98}_{44}\text{Ru}$
- $^{99}_{44} \text{Ru}$
- $^{100}_{44} \text{Mo}$, $^{100}_{44}\text{Ru}$

```python
from matplotlib.pyplot import subplots

MH = 1.007825 #u
Mn = 1.00866491588 #u

def MeV_to_J(energy):
    return energy*1.609e-13

def epsilon_5(A, Z, a):
    if A % 2 == 0 and Z % 2 == 0:
        return a[5]/A**(3/4)
    elif A % 2 == 1 and Z % 2 == 1:
        return -a[5]/A**(3/4)
    else:
        return 0   

c = 3e8

a = {
    1: 15.76, #MeV
    2: 17.81, #MeV
    3: 0.7105, #MeV
    4: 98.80, #MeV
    5: 39.0 #MeV
}

M = lambda A, Z, a: Z*MH + (A-Z)*Mn - (a[1]*A - a[2]*A**(2/3) - a[3]*Z**2/A**(1/3) - a[4]*(A/2 - Z)**2 /A + epsilon_5(A, Z, a))/c**2
```
