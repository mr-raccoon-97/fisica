### Estructura nuclear y radiactividad

El núcleo se encuentra en el centro del átomo ocupando solo una $10^{-15}$ de su volumen.

Dentro del núcleo, hay $Z$ cargas positivas. Para evitar que estas salgan disparadas, el núcleo debe proveer una fuerza nuclear que supere a la repulsión colombiana entre estos.

Hay muchas similitudes entre la estructura atómica y la estructura nuclear. El núcleo está sujeto a la física cuántica. Este tiene estados fundamentales y excitados y puede emitir fotones durante las transiciones entre estados, por lo que pueden ser caracterizados por su momento angular.

Hay sin embargo dos grandes diferencias entre el estudio de las propiedades atómicas y nucleares. En la física atómica, los electrones experimentan una fuerza provista por un agente externo, el núcleo. En la física nuclear no existe tal agente y en contraste con la física atómica, en la cual podíamos considerar la interacción entre electrones como una perturbación, en la física nuclear las acciones de los constituyentes nucleares son lo que proveen la fuerza, por lo que no pueden ser ignorados.

La segunda diferencia entre la física atómica y la física nuclear es que no se puede escribir la fuerza nuclear de una forma simple como la fuerza de Coulomb. No existe una expresión analítica cerrada para describir la interacción entre los constituyentes.

### Constituyentes nucleares

Los trabajos de Rutherford y Bhor demostraron que las cargas positivas de un átomo, se encuentran confinadas en una región muy pequeña en el centro de este. Un núcleo con número atómico $Z$ tiene una carga de $Ze^+$, y del núcleo proviene el $99.9\%$ de la masa. Se sabe también que las masas son casi enteros si se los cuenta en términos de la masa del protón. Denominamos a estos enteros como $A$ el número de masa.

La diferencia entre el número de masa y el número atómico da como resultado el número de neutrones en un átomo.

Al tener propiedades nucleares similares, el protón y el neutrón, son ambos denominados nucleones. 

Las propiedades químicas de un elemento se dependen del numero $A$ y no del número $A$, ya que estas están asociadas generalmente a las cargas.

### El modelo de gas de Fermi

Los nucleones son fermiones y por lo tanto están sujetos al principio de exclusión de Pauli. Por lo tanto, estos se encuentran bajos las mismas restricciones que los electrones en un átomo.

El principio de exclusión nos permite llenal los niveles con un máximo de dos nucleones, uno con espín hacia arriba y otro con espín hacia abajo. En la siguiente figura se muestra la distribución de los nucleones en el núcleo para diferentes números atómicos $Z$ y números de masa $A$.




Las configuraciones más cercanas a $Z=N$ tienden a ser más estables.

Podemos emplear la antisimétrica de los fermiones para justificar la expresión de la energía de ligadura $E_b = \sum_i M_i c^2 - Mc^2$. En donde $M_i$ es la masa de cada nucleón y $M$ la masa del núcleo. Consideremos al núcleo como una colección grande de nucleones moviéndose libremente en una esfera cerrada definida por el volumen nuclear, y describimos al sistema como un gas de Fermi degenerado a temperatura cero, en los cuales los nucleones ocupan las energías más bajas consistentemente con el principio de exclusión de Pauli.

El protón y el neutrón son distinguibles entre sí, de modo que la estadística de Fermi-Dirá es aplicada a estos de forma separada con la ayuda de la siguiente figura, en la cual se introduce una barrera de energía de potencial constante para el movimiento libre de cada partícula individual.

Dado que $A$ es grande, podemos incorporar el efecto culombiano repulsivo para los protones y emplear barreras de potencial de diferente altura para los protones y los neutrones. Un gran número de niveles de energía es asumido en ambos casos y todos se presumen ocupados por los gases de neutrones y protones hasta los niveles de energía de Fermi indicados.

Nótese que la energía de Fermi y el número de partículas son más grandes en el caso de los neutrones debido a que el efecto culombiano eleva la barrera de potencial en el caso de los protones.

Habíamos visto que el volumen del núcleo se encuentra determinado por el número de masa $A$ de la forma: 

$$ V = \frac{3}{4}\pi R_0 A^3$$

Expresamos el número de nucleones en terminos de las energias de Fermi:

$$ Z = \frac{(2M_p)^{3/2}}{3\pi^2\hbar^3}V \epsilon_{F_p}^{3/2} \qquad N = \frac{(2M_n)^{3/2}}{3\pi^2\hbar^3}V \epsilon_{F_n}^{3/2}$$

Y luego despejamos las energías de Fermi:

$$ \epsilon_{F_p} = \frac{\hbar^2}{2M_pR^2} \left(  \frac{9\pi}{4} \frac{N}{A} \right) \qquad \epsilon_{F_n} = \frac{\hbar^2}{2M_nR^2} \left(  \frac{9\pi}{4} \frac{N}{A} \right) $$

Luego, la energia total en función de las energías de Fermi y el numero de partículas es:

$$ E_Z = \frac{3}{5} \epsilon_{F_p} Z = \frac{3}{10} \frac{\hbar^2}{MR_0^2}Z\left( \frac{9\pi}{4} \frac{Z}{A} \right) ^{2/3} = \frac{3}{10} \left( \frac{9 \pi}{4} \right)^{2/3} \frac{\hbar^2}{MR_0^2} A \left( \frac{Z}{A} \right)^{5/3}$$

$$ E_N = \frac{3}{10} \left( \frac{9 \pi}{4} \right)^{2/3} \frac{\hbar^2}{MR_0^2} N \left( \frac{Z}{N} \right)^{5/3}$$

Nos interesa la simetria de energia del nucleo debido a la desviación del numero de protones y neutrones de la igualdad $Z=N = A/2$. Introducimos una nueva variable $\zeta$ tal que:

$$ N = \frac{A}{2} + \zeta \qquad Z = \frac{A}{2} - \zeta$$

Puede demostrarse, utilizando expansiones binomiales que:

$$ E_Z + E_N = \frac{3}{40} (9\pi)^{2/3} \frac{\hbar^2}{MR_0^2} \left( A + \frac{20}{9} \frac{\zeta^2}{A} \right)$$

A segundo orden de $\zeta$. Si reemplazamos $\zeta$ por $A/2 - N$, la contribución a segundo orden, del término de simetría es:

$$ \frac{(9\pi)^{2/3}}{6} \frac{\hbar^2}{MR_0^2} \frac{(A/2 - Z)^2}{A} $$

### Interacciones nucleón-nucleón

Hasta ahora los modelos discutidos, tenían que ver con las propiedades globales de los núcleos y no con los detalles de la estructura nuclear.

El problema fundamental de la física nuclear es la determinación de la fuerza entre dos nucleones. Se puede estudiar la interacción entre nucleones a través de la dipersión sobre un protón como blanco. 

El deuterio es el único núcleo con un número másico $A=2$ y es el único sistema $pn$, no hay contra partes $pp$ o $nn$. 

Las características del deuterio son:

- Energía de ligadura $ E_b(^2H) = 2.225 \text{MeV}$
- Espín nuclear y paridad $i^p = 1^+$
- Momento dipolar magnético $\mu_d = 0.8574 \mu_N$
- Momento cuadrupolar eléctrico $Q_d = 2.85 \times 10^{-3} e \cdot \text{barn}$
- Radio nuclear $R_d = 2.1\text{fm}$

Estas caracteristicas se obtienen de diferentes fuentes experimentales. Una forma de medir la energia de ligadura del deuterio a través de la medición de los rayos $\gamma$ emitidos en la reacción de formación:

$$ n + p \rightarrow d + \gamma$$

El espín nuclear y el momento diploar magnetico se obtienen de la medición de la estructura hiper-fina y de experimentos de resonancia magnetica. 

El momento cuadrupolar se obtiene de la aplicación de la tecnica resonancia de rayos y se interpreta como un indicador la forma nuclear. 

El radio nuclear se obtiene de la medición de la sección eficaz de dispersión de electrones.

### Momento angular

Escribimos el momento angular del deuterio como el momento angular del sistema neutrón-protón:

$$ \vec{I} = \vec{L} + \vec{S_p} + \vec{S_n}$$

En donde que $S_p$ y $S_n$ son espines nucleares de spin $\frac{1}{2}$. Se puede generazliar esta expresión para cualquier núcleo con $Z$ protones y $N$ neutrones:

$$ \vec{I} = \sum_{k=1}^Z ( \vec{L}_{p_k} + \vec{S}_{p_k}) + \sum_{k=1}^N ( \vec{L}_{n_k} + \vec{S}_{n_k})$$	

Se sabe de antemano que el vector $\vec{I}$ y su número cuántico $i$ están cuantizados, tales que los autoestados y autoenergias del operador $\hat{I}^2$ y $\hat{I}_z$ son:

$$ \hat{I}^2 |i,m_i \rangle = i(i+1) \hbar^2 |i,m_i \rangle \qquad i = i_{mín}, i_{mín}+1, ..., i_{máx}$$


$$ \hat{I}_z |i,m_i \rangle = m_i \hbar |i,m_i $$

En donde: 
- $m_i = -i, -i+1, ..., i-1, i$ si $A$ es par.
- $m_i = -i, -i+\frac{1}{2}, ..., i-\frac{1}{2}, i$ si $A$ es impar.

Esta teoría es el analogo exacto para la teoría del momento angular en átomos multielectrónicos.

### Modelo nuclear de capas



Los estados de un nucleón individual se encuentran adoptando un modelo de campo central que describe las interacciones entre el nucleón y las otras $A-1$ partículas. El principio de exclusión de Pauli, gobierna a los neutrones y protones por separado y controla los estados de ocupación para los niveles de energía individual de un nucleón para todos los $A$ nucleones.

Supongamos un potencial central $V(r)$ para cada nucleón, de forma tal que podemos aplicar las propiedades de cuantización del momento angular de los estados a cada partícula individual. Esto nos permite etiquetar los estados por los números cuánticos de una sola partícula $n, l, m_l, m_s$. Luego las autofunciones para un nucleón individual son:

$$ \psi_{nlm_lm_s} = R_{nl}(r) Y_{lm_l}(\theta, \phi) \chi_{m_s}$$

La función radial $R_{nl}(r)$ y su energía asociada $E_{nl}$ dependen del potencial elegido $V(r)$. La ecuación de Schrödinger radial para un nucleón individual es:

$$ \frac{d^2u_{nl}}{dr^2} (r R_{nl}) + \frac{2\mu}{\hbar^2} [E_{nl} - V_{efectivo}(r)] (r R_{nl}) = 0 $$

En donde el potencial efectivo incluye el término centrífugo:

$$ V_{efectivo}(r) = V(r) + \frac{\hbar^2}{2\mu} \frac{l(l+1)}{r^2}$$

Un candidato posible al potencial $V(r)$ es:

$$ V(r) = - \frac{V_0}{1 + e^{(r-R)/a}} \qquad R = R_0 A^{1/3}$$

```python
#Grafico del potencial V(r)

from math import exp
from matplotlib.pyplot import subplots

V = lambda r, V0, a, R: -V0/(1 + exp((r - R)/a))

figure, ax = subplots()
r = [i/100 for i in range(0, 100)]
V0 = 1
a = 0.1
R = 0.5
ax.plot(r, [V(i, V0, a, R) for i in r])
ax.grid()
```

Se puede aproximar el potencial $V(r)$ como un pozo cuadrado de profundidad $V_0$ que admite soluciones exactas y luego mejorar la aproximación numéricamente si hace falta.

La descripción de los niveles de energía del nucleón individual aplica al caso en el que el nucleón sea un neutrón. Para el caso del protón se debe añadir el término de interacción culombiana con los otros $A-1$ protones.

El modelo de capas emerge de la ecuación de Schrödinger radial planteada con solo índices $n$ y $l$. La degeneración existe en todos los niveles debido a que la energía no depende de $m_l$ y $m_s$. Luego a cada valor de $E_{nl}$ le corresponden $2(2l+1)$ estados degenerados.

Debido a esto último, el modelo de capas no es suficiente para describir la estructura nuclear. Se debe considerar la interacción espín-órbita nuclear. Esta interacción causa una separación en los niveles de energía $E_{nl}$ para todos los valores de $l \neq 0$. La separación ocurre porque la interacción atómica contribuye a un cambio de energía distinto para los dos espines de un número cuántico $l$ dado.

Esta interacción se puede describir en términos del potencial central $V(r)$ como:

$$ V_{SL} = - \frac{a^2_{SL}}{r} \frac{dV}{dr} \vec{L} \cdot \vec{S}$$

En donde $a_{SL}$ es una constante que determina la separación deseada. La interacción espín-órbita es la responsable de la separación de los niveles de energía en el modelo de capas, como muestra la siguiente figura.

En la figura anterior se observa como la interacción espín-órbita separa los niveles de energía $E_{nl}$ en pares $E_{nlj}$, con $j = l \pm 1/2$.
