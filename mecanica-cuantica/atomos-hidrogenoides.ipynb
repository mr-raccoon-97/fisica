### El tomo de hidrgeno

Un electrn interactuando con un protn en un tomo de hidrgeno puede describirse como un problema de dos cuerpos, separando el movimiento del centro de masa del movimiento relativo entre ellas. Como la masa del protn es 1835 veces mayor que la del electrn, el centro de masa prcticamente coincide con la posicin del protn, mientras que la masa reducida del sistema es casi idntica a la del electrn. 

Por este motivo, lo habitual es ignorar el movimiento del tomo como un todo y solo considerar el electrn bajo la accin de un potencial central. El campo coulombiano sobre el electrn es $V(r) =e^2/r$ , donde $e$ es la carga del electrn y $r$ , la distancia entre ambas partculas. La ecuacin radial reducida se escribe entonces como:

$$ (-\frac{\hbar^2}{2\mu} + \frac{\hbar^2 l(l+1)}{2\mu r^2} - \frac{e^2}{r}) u(r) = E $$

Las primeras funciones radiales para un tomo de hidrgeno son $\psi_{nlm}(r,\theta,\phi) = R_{nl}(r)Y_{lm}(\theta,\phi)$, donde $R_{nl}(r)$ es la funcin de onda radial y $Y_{lm}(\theta,\phi)$ es la funcin de onda angular.
