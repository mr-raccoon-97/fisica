### El átomo de hidrógeno

Un electrón interactuando con un protón en un átomo de hidrógeno puede describirse como un problema de dos cuerpos, separando el movimiento del centro de masa del movimiento relativo entre ellas. Como la masa del protón es 1835 veces mayor que la del electrón, el centro de masa prácticamente coincide con la posición del protón, mientras que la masa reducida del sistema es casi idéntica a la del electrón. 

Por este motivo, lo habitual es ignorar el movimiento del átomo como un todo y solo considerar el electrón bajo la acción de un potencial central. El campo coulombiano sobre el electrón es $V(r) =−e^2/r$ , donde $e$ es la carga del electrón y $r$ , la distancia entre ambas partículas. La ecuación radial reducida se escribe entonces como:

$$ (-\frac{\hbar^2}{2\mu} + \frac{\hbar^2 l(l+1)}{2\mu r^2} - \frac{e^2}{r}) u(r) = E $$

Las primeras funciones radiales para un átomo de hidrógeno son $\psi_{nlm}(r,\theta,\phi) = R_{nl}(r)Y_{lm}(\theta,\phi)$, donde $R_{nl}(r)$ es la función de onda radial y $Y_{lm}(\theta,\phi)$ es la función de onda angular.
