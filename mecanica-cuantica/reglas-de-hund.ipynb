### Reglas de Hund

En un átomo multi-electrónico, puede asumirse que el potencial promedio al que está sometido cada electrón mantiene simetría esférica, es decir se trata de un potencial central.

De manera similar al caso del átomo hidrogenoide, se tendrán autoestados $\ket{n,l,m_l,m_s}$. Estos se organizarán de la siguiente forma:

* Los electrones atómicos se ordenan en una estructura de capas, cada una de las cuales se caracteriza mediante el número cuántico principal o radial $n$.

* Estas capas tienen subcapas especificadas por los números cuánticos orbitales $l$, que representa los posibles autovalores para el módulo del momento angular orbital $\hat{L}^2$.

* A su vez, estas se dividen en subcapas más pequeñas según el valor de las posibles proyecciones $m_l$, es dedcir autovalores del operador $\hat{L}_z$. Estas subcapas se denominan orbitales atómicos.

* Finalmente, como los electrones son fermiones, deben cumplir el principio de exclusión de Pauli, lo que implica que no pueden ocupar el mismo estado cuántico. Por lo tanto, cada orbital atómico puede contener a lo sumo dos electrones, uno con espín $\ket{\uparrow}$ y otro con espín $\ket{\downarrow}$.


Las soluciones que se logran para átomos multi-electrónicos repiten el esquema de desdoblamientos que se habían visto para los átomos hidrogenoides, cada capa tiene $n$ capas con $l=0,1,2,...,n-1$ las cuales admiten $2l+1$ proyecciones y se identifican con las letras $s, p, d, f, g$ etc.

Para un dado $n$, un estado $s$ ($l=0$) tiene solo un orbital, los estados $p$ ($l=1$) incluyen tres orbitales ($m_l = -1, 0, 1$), etc. Así hablamos de los orbitales $1s, 2s, 2p, 3s, 3p, 3d, ...$ etc.

Los orbitales se van llenando en orden de energías crecientes y la distribución de los electrones en los diferentes orbitales se denominan **configuraciones electrónicas**.


El ordenamiento de las energías individuales sigue en general la siguiente regla: escribiendo columnas para
los orbitales permitidos en cada capa, en orden creciente de $n$ (hacia la derecha) y de $l$ (hacia abajo), las energías crecientes se abarcan recorriendo las diagonales de abajo hacia arriba y de izquierda a derecha.

La forma de colocar un par de electrones en los orbitales será entonces como sigue:

Al resolver el problema del átomo multielectródico, se encuentra que la energía de cada configuración depende del valor del momento angular total del sistema $\vec{J}$, a traves de la relación:

$$ \vec{J} = \sum_{j=1}^Z \vec{L}_j + \vec{S}_j $$

En donde se verifican las restricciones:

$$ | L - S | \leq J \leq L + S \qquad \text{y} \quad -J \leq M_J \leq J $$

Con $L$ no necesariamente mayor que $S$, en el caso de $L > S$ hay $2S + 1$ valores posibles para $J$, estos son $J = L-S, L-S+1, ..., L+S-1, L+S$, mientras que si $L < S$ hay $2L + 1$ valores posibles para $J$, estos son $J = S-L, S-L+1, ..., S+L-1, S+L$. 

Cada valor de $J$ admite $2J + 1$ proyecciones posibles, de modo que la energía estará degenerada, pues no depende de los valores que pueda tomar $M$. Se introduce entonces la notación espectroscópica:

$$ ^{2S+1}L_J $$

Y al igual que para los estados individuales, se asocian a los estados $L$ las letras $S, P, D, F, G, ...$ y a los valores de $J$ los números $0, 1, 2, 3, 4, ...$. 

Como se buscan, las configuraciones electrónicas del estado fundamental, hay que hallar el valor resultante de $J$ tal que la energía sea mínima. Para ello, se sigue la regla de Hund:

- $S$ es máximo.
- $L$ es máximo compatible con $S$ máximo.
- En subcapasa $nl$ llenas hasta la mitad, $J=|L-S|$, en caso contrario, cuando hay más de la mitad de los electrones $J=L+S$.
